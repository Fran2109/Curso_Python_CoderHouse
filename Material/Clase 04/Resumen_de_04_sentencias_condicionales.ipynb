## Sentencias condicionales

Las sentencias condicionales son instrucciones que nos permiten ejecutar un bloque de codigo despues de haber evaluado una expresion logica, conocida como **CONDICION**

En Python, existen tres palabras claves:
* if
* elif
* else



## Sentencia IF

`if` es la instruccion principal. Nos permite controlar el flujo del programa y dividir la ejecucion

Cuando invocamos esta instruccion, le decimos a Python que queremos evaluar una condicion y en base a ello, ejecutar instrucciones.

Siempre? No, solo cuando la condicion de como resultado el valor True.

Si el resultado de la condicion es False, simplemente seguira con el resto del programa sin ejecutar el bloque de instrucciones "dentro" del `if`

````python
edad = int(input("Introduzca su edad: "))

if edad >= 18:
  print("Es mayor de edad")
````

````python
edad = input("Introduzca su edad: ")

if True:
  print("Esto siempre se va a ejecutar, ya que en lugar de la condicion, le pasamos un literal True")
````

En las condiciones podemos utilizar cualquier operador aritmetico y logico que hayamos visto hasta el momento

````python
usuario_valido = "admin"
dominio_valido = "example.com"
contraseña_valida = "admin123"

correo = input("Ingrese su correo: ")
contraseña = input("Ingrese su contraseña: ")

if not len(correo) < 6 and correo == (usuario_valido+"@"+dominio_valido) and contraseña == contraseña_valida:
  print("Inicio de sesion exitoso! Bienvenid@ admin")
````

## Identacion:

Es importante notar que por debajo de la instruccion `if` estamos declarando instrucciones con una separacion de la donde inicia la sentencia.

```python
if condicion:
  print("Imprimo este mensaje porque la condicion es True")
```

Si lo escribieramos de esta forma, el programa fallaria:
```python
if condicion:
print("Imprimo este mensaje porque la condicion es True")
```

Debemos ser consistentes con la identacion. Si establecemos 2 espacios, debemos ir anidado las instrucciones con 2 espacios y no con 3, 4 o inclusive con TAB

```python
if True:
  print("Inicio con 2 espacios")
   print("Aca hay 3 espacios y ya marca un error")
```

## Condiciones anidadas

Dado que los `if` son instrucciones, podemos introducirlas dentro de otros `if` y seguir evaluando cosas.

Podemos escribir tantos `if` como necesitemos

````python
# Ejemplo: Si el cliente pago con EFECTIVO una suma inferior a $1500, 
# se le aplicara un descuento del 10% sobre el monto de la compra.
# Si pago con tarjeta por un monto superior o igual a $3000, apicamos un recargo del 
# 15% sobre el monto de la compra

monto_compra = float(input("Ingrese el valor total de la compra: $"))
metodo_pago = input("Ingrese el metodo de pago (E: Efectivo, T: Tarjeta): ")

descuento = 0
recargo = 0

if metodo_pago == "E":
  if monto_compra < 1500:
    descuento = monto_compra * 0.10

if metodo_pago == "T":
  if monto_compra >= 3000:
    recargo = monto_compra * 0.15

if descuento > 0:
  print("Se ha aplicado un descuento de $", descuento,". Total a pagar: $", (monto_compra-descuento))
if recargo > 0:
  print("Se ha aplicado un recargo de $", recargo,". Total a pagar: $", (monto_compra+recargo))
if descuento == 0 and recargo == 0:
  print("Total a pagar: $", monto_compra)
````

## Sentencia ELSE

La sentencia `else` SIEMPRE va acompañada de un `if` antes.

A diferencia del `if`, else ejecuta un bloque de codigo cuando la condicion evaluada sea False.

````python
edad = int(input("Introduzca su edad: "))

if edad >= 18:
  print("Es mayor de edad")
else:
  print("Es menor de edad")
````

````python
edad = int(input("Introduzca su edad: "))

if False:
  print("esto jamas se va a mostrar")
else:
  print("Este mensaje siempre se va a mostrar, independientemente del numero que coloques")
````

## Sentencia ELIF

La sentencia `elif` es la mas curiosa de todas. Es similar a un `else` pero le acompaña una condicion mas.

Se utiliza despues de un `if` y antes de un `else`.


````python
# Calculadora de IMC: 
# El IMC es la relación entre la estatura de una persona y su peso

# Formula: peso(Kg) / (altura(mts)*altura(mts))

peso = float(input("Ingrese su peso en Kg.: "))
altura = float(input("Ingrese su altura en Mts.: "))

imc = peso / altura**2

# Redondeamos digitos
# La funcion round es util para limitar la cantidad de digitos despues del punto decimal
# Simplemente le indicamos en primer lugar la variable, y luego la cantidad de 
# digitos que queremos posteriores al punto decimal.
imc = round(imc, 2)

if imc < 18.5:
  print("Resultado",imc, ". Relacion peso-altura: Bajo peso")
elif imc >= 18.5 and imc < 25:
  print("Resultado",imc, ". Relacion peso-altura: Normal")
elif imc >= 25 and imc < 30:
  print("Resultado",imc, ". Relacion peso-altura: Sobrepeso")
elif imc >= 30 and imc < 35:
  print("Resultado",imc, ". Relacion peso-altura: Obesidad I")
elif imc >= 35 and imc < 40:
  print("Resultado",imc, ". Relacion peso-altura: Obesidad II")
elif imc >= 40 and imc < 50:
  print("Resultado",imc, ". Relacion peso-altura: Obesidad III")
else:
  print("Resultado",imc, ". Relacion peso-altura: Obesidad IV")
````

Como podemos observar, la sentencia `elif` nos permite seguir evaluando otras condiciones pese a que la condicion del `if` dio False.

Se pudo haber hecho solo con `if` y `else`. Si, pero el codigo se hubiera visto asi:


````python
# Calculadora de IMC: 
# El IMC es la relación entre la estatura de una persona y su peso

# Formula: peso(Kg) / (altura(mts)*altura(mts))

peso = float(input("Ingrese su peso en Kg.: "))
altura = float(input("Ingrese su altura en Mts.: "))

imc = peso / altura**2

# Redondeamos digitos
# La funcion round es util para limitar la cantidad de digitos despues del punto decimal
# Simplemente le indicamos en primer lugar la variable, y luego la cantidad de 
# digitos que queremos posteriores al punto decimal.
imc = round(imc, 2)

if imc < 18.5:
  print("Resultado",imc, ". Relacion peso-altura: Bajo peso")
else:
  if imc >= 18.5 and imc < 25:
    print("Resultado",imc, ". Relacion peso-altura: Normal")
  else:
    if imc >= 25 and imc < 30:
      print("Resultado",imc, ". Relacion peso-altura: Sobrepeso")
    else:
      if imc >= 30 and imc < 35:
        print("Resultado",imc, ". Relacion peso-altura: Obesidad I")
      else:
        if imc >= 35 and imc < 40:
          print("Resultado",imc, ". Relacion peso-altura: Obesidad II")
        else:
          if imc >= 40 and imc < 50:
            print("Resultado",imc, ". Relacion peso-altura: Obesidad III")
          else:
            print("Resultado",imc, ". Relacion peso-altura: Obesidad IV")
````

Es un codigo que funciona igual PERO que es dificil de seguir.

Tambien podrias pensar "Ok, pero puedo hacerlo con puros if"

La repuesta es SI! Pero tendrias que pensar mucho en las condiciones. Mira el siguiente ejemplo:



````python
calificacion = int(input("Ingrese su calificacion del 1 al 10: "))

if calificacion >= 9:
  print("Sobresaliente")
if calificacion >= 7:
  print("Muy bien")
if calificacion >= 6:
  print("Bien")
if calificacion >= 4:
  print("Regular")
else:
  print("Insuficiente")
````

Al introducir el numero 10, vemos que todos los print se han ejecutado.

Esto se debe a que cada `if` se evalua de forma independiente.

El `else` que se encuentra al final solo esta atado al ultimo `if` y no a los anteriores, asi que si pensamos mal la condicion de esa sentencia, podriamos haberlo ejecutado accidentalmente tambien.

Si queremos que funcione, debemos de pensar en una condicion que solo de verdadero en un solo `if`, pero asi y todo tendriamos problemas:

````python
calificacion = int(input("Ingrese su calificacion del 1 al 10: "))

if calificacion >= 9:
  print("Sobresaliente")
if calificacion >= 7 and calificacion < 9:
  print("Muy bien")
if calificacion >= 6 and calificacion < 7:
  print("Bien")
if calificacion >= 4 and calificacion < 6:
  print("Regular")
else:
  print("Insuficiente")
````

Al introducir una nota cualquiera, como 8, vemos que se imprime el mensaje correspondiente PERO TAMBIEN se imprime el mensaje ubicado dentro del `else`

Podemos seguir agregando mas condiciones, quitar ese else y reemplazarlo con un `if`, etc.

Pero para que hariamos todo eso si con algunos `elif` lo solucionamos?

````python
calificacion = int(input("Ingrese su calificacion del 1 al 10: "))

if calificacion >= 9:
  print("Sobresaliente")
elif calificacion >= 7:
  print("Muy bien")
elif calificacion >= 6:
  print("Bien")
elif calificacion >= 4:
  print("Regular")
else:
  print("Insuficiente")
````

Pero como funciono esto y no lo de antes?

Porque Python evalua condicion por condicion hasta encontrar alguna verdadera.

Si introducimos el 10: 
* Python comenzara con el `if`
* Evalua la condicion. Da un resultado Verdadero
* Ejecuta el bloque de codigo
* Cuando el bloque se termina, sigue  con el resto del programa. No evalua ninguna otra sentencia `elif` ni tampoco va por un `else`

Si introducimos el 6:
* Python ira por el `if`
* Evalua la condicion. Da un resultado Falso
* La siguiente sentencia que encuentra es un `elif`
* Evalua la condicion. Da un resultado Falso
* Encuentra otro `elif`.
* Evalua la condicion. Da un resultado Verdadero
* Ejecuta el bloque de codigo
* Cuando el bloque se termina, sigue  con el resto del programa.

Creo que ya entendiste como funciona 😀

# IF en profundidad

Todo valor que reciba la instruccion `if`, es convertida a `bool`.

Te estaras preguntando, "..._para que necesito eso si mis condiciones ya son booleanas?_"

Esto no cambia nada la forma en la que ya trabajamos, pero si abre las puertas a nuevas expresiones mas simples y cortas.

Mira este ejemplo:


````python
usuario = input("Ingrese su usuario: ")

if not usuario:
  print("Debe ingresar un nombre de usuario")
else:
  print("Bienvenid@", usuario)
````

Ahora tu mente debe estar confundida

* _por que coloco "usuario" si no es un valor booleano o una condicion? Se supone que el IF solo evalua True o False_

Si eso es cierto y, como te dije antes, no cambio nada.

Mas arriba lo dije: Python toma todo valor que recibe el if y lo convierte en booleano

Te acordas de `False == 0`?

Bueno, pasa algo similar con otros tipos de datos!!

````python
# Numeros
numero_positivo = 10
numero_negativo = -10.5
numero_cero = 0

print("Numero", numero_positivo, "equivale a ", bool(numero_positivo))
print("Numero", numero_negativo, "equivale a ", bool(numero_negativo))
print("Numero", numero_cero,     "equivale a ", bool(numero_cero))
print("-"*30)


# String
string_no_vacio = "Hola mundo"
string_vacio = ""

print("String", string_no_vacio, "equivale a ", bool(string_no_vacio))
print("String vacio:", string_vacio, "equivale a ", bool(string_vacio))
print("-"*30)

# Listas
lista_no_vacia = [1,2,3]
lista_vacia = []

print("Lista", lista_no_vacia, "equivale a ", bool(lista_no_vacia))
print("Lista", lista_vacia,    "equivale a ", bool(lista_vacia))
print("-"*30)

# Tuplas
tupla_no_vacia = (1,2,3)
tupla_vacia = tuple()

print("Tupla", tupla_no_vacia, "equivale a ", bool(tupla_no_vacia))
print("Tupla", tupla_vacia,    "equivale a ", bool(tupla_vacia))
````
