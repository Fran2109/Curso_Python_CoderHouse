# Persistencia

Todos los programas que hemos realizado hasta hoy, mantienen la informacion de forma temporal. Cuando finalizan, todo se pierde 💨 y debemos iniciar de cero.

Existira alguna forma de hacer que la informacion perdure? Si, la hay

La persistencia en programacion, es la accion de preservar (guardar) informacion y recuperarla (leerla).

Existen dos formas de persistir los datos

## Bases de datos

Las bases de datos son la mejor solucion para esto. Son sistemas complejos que cuentan con mecanismos para manipular los datos que se encuentran en ellas. Operaciones como guardar, leer, actualizar y borrar se hacen mas sencillas, rapidas y seguras cuando lo hacemos con esta forma de persistencia


Pero tenemos un problema: requieren un esfuerzo mayor.

Debemos de tener un minimo de conocimiento acerca de como funcionan, como instalarlas, como trabajar con los datos, etc.

## Archivos

Los archivos son la forma mas comun y facil de persistir informacion cuando solo queremos guardar y recuperar datos.

Un archivo no es mas que una secuencia de bytes almacenados en un dispositivo, identificado bajo un nombre y colocado en alguna carpeta.

Existen dos tipos de archivos:

+ Binarios: como su nombre lo indica, son archivos que contienen informacion binaria, ilegible para el ser humano pero procesable por nuestra computadora. Ejemplos: archivos de audio, video, programas .exe, etc.

+ Texto: con el nombre ya te debes de imaginar que contiene. La informacion en este tipo de archivos es legible por nosotros puesto que no son mas que caracteres ASCII. De estos tenemos un monto: archivos txt, json, xml, html, etc.


Si abro cualquiera de estos archivos en algun editor (como el bloc de notas) puedo leer palabras, instrucciones, etc. Esto no pasa con un archivo binario

## Escribir informacion en archivos

Python nos ofrece una funcion para abrir archivos y trabajar con el contenido del mismo: `open`

Esta funcion requiere de dos cosas: la ruta donde se encuentra el archivo y el modo de apertura.


Como estamos en Colabs, si damos click derecho sobre alguna carpeta y seleccionamos "Copiar Ruta", nos copiara una cadena con un formato particular.

Esa ruta es conocida como "ruta absoluta".

En el caso de Windows las rutas absolutas se ven asi: `C:\Usuarios\pepe\Documentos`

En Linux se ven asi: `/home/pepe/Documentos`


La funcion `open` permite que el string que le pasamos como "ruta" sea del tipo "ruta relativa" o "ruta absoluta"

Abajo las diferencias 🔽





### RUTAS ABSOLUTAS VS. RUTAS RELATIVAS

En las rutas absolutas tenemos que indicar todo el "camino" que debe recorrer Python para llegar al archivo.

`Ruta Absoluta: C:\Users\usuario\Documents\carpeta1\carpeta2\carpeta3\archivo.txt`

Como podes observar, le decimos que inicio desde el disco "C", acceda a la carpeta "Users", dentro de esa acceda a la carpeta "usuario", y asi sucesivamente hasta llegar al archivo.

---

En cambio con las rutas relativas, Python toma como referencia el lugar donde se encuentra el archivo python (eso significa el `.\` ) que se esta ejecutando y se va moviendo por las carpetas segun indique la ruta.

En este caso le estamos diciendo que desde donde esta, debe acceder a la carpeta llamada "carpeta" y dentro esta el archivo que debe abrir.

`Ruta Relativa: .\carpeta\archivo.txt`

### MODOS DE APERTURA

Los archivos cuentan con distintos modos de apertura.

+ Modo `"r"`: abre el archivo para su lectura
+ Modo `"r+"`: abre el archivo tanto para lectura como para escritura
+ Modo `"w"` : abre un archivo con el único propósito de escribir.
+ Modo `"w+"`: abre el archivo tanto para lectura como para escritura
+ Modo `"a"` : abre un archivo con el único propósito de escribir.
+ Modo `"a+"`: abre el archivo tanto para lectura como para escritura
+ Modo `"x"` : abre un archivo con el único propósito de escribir.
+ Modo `"x+"` : abre el archivo tanto para lectura como para escritura

Si bien puede parecer que hacen lo mismo, las diferencias entre los distintos modos de apertua son significativas

````python
ruta = "/content/drive/MyDrive/coder_apunte.txt"

archivo = open(ruta,  "w")
````

Una vez que tenemos abierto nuestro archivo, es momento de escribir contenido en el.

Como lo hemos abierto en el modo de "solo escritura", Python nos permite ejecutar solo funciones que cumplan ese proposito.

La funcion para este caso, es la de `write`. Esta funcion recibe un string. Solo uno.

Es importante ejecutar el metodo `close` antes de que el programa termine, ya que los cambios que hayamos hecho sobre un archivo, no se veran reflejados  hasta que esa funcion sea ejecutada.

````python

cadena = "Esto es una cadena de prueba que estamos agregando a nuestro archivo :)"

archivo.write(cadena)

archivo.close()
````

## Lectura de archivos

Excelente! Hasta ahora hemos visto como podemos guardar informacion. Solo nos resta saber como recuperarla.

La respuesta es sencilla: solo cambiamos el modo de apertura y el metodo que ejecutamos.

````python
ruta = "/content/drive/MyDrive/heroes_marvel.txt"

archivo = open(ruta,  "r")
````

Ahora que lo abrimos bajo ese modo, tenemos disponibles tres funciones:

+ read: devuelve en un string todo el contenido del archivo
+ readline: cada vez que ejecutamos esta funcion, devuelve una linea nueva
+ readlines: devuelve una lista donde cada item es una linea en el archivo

````python
# Metodo read

contenido = archivo.read()

print(f"Contenido del archivo: {contenido}")
print(f"Tipo de dato de contenido: {type(contenido)}")

archivo.close()
````

````python
# Metodo readline

ruta = "/content/drive/MyDrive/heroes_marvel.txt"

archivo = open(ruta,  "r")

contenido = archivo.readline()

while contenido:
  print(f"Contenido del archivo: {contenido}")
  print(f"Tipo de dato de contenido: {type(contenido)}")
  print("=========================")
  contenido = archivo.readline()

archivo.close()
````

````python
# Metodo readlines

ruta = "/content/drive/MyDrive/heroes_marvel.txt"

archivo = open(ruta,  "r")

contenido = archivo.readlines()

print(f"Contenido del archivo: {contenido}")
print(f"Tipo de dato de contenido: {type(contenido)}")

archivo.close()
````

Hasta aca parece ir todo bien, pero tenemos un problema del que no nos percatamos.

Mira lo siguente:

````python
# Metodo readlines

ruta = "/content/drive/MyDrive/heroes_marvel.txt"

archivo = open(ruta,  "r")

# Me traigo el contenido por primera vez
contenido = archivo.readlines()

print(f"Contenido del archivo: {contenido}")
print(f"Tipo de dato de contenido: {type(contenido)}")


# Me traigo el contenido por segunda vez
contenido = archivo.readlines()

print(f"Contenido del archivo: {contenido}")
print(f"Tipo de dato de contenido: {type(contenido)}")

archivo.close()
````

Me trajo una lista vacia?

Si. Esto se debe a que los archivos cuentan con unos apuntadores que le indican a Python donde debe serguir leyendo. 

Cada vez que leemos datos (con cualquiera de los metodos), el apuntador se mueve hacia adelante hasta llegar al final.

Una vez que llego al final, Python no puede seguir leyendo.

Por suerte para nosotros, contamos con un metodo que me permite mover ese apuntador a gusto.

El metodo se llama `seek` y recibe una valor numerico.

Ese valor le indica cuantos bytes (o caracteres, como quieras decirle) debe moverse desde el inicio (siempre toma como referencia el inicio).

Una vez modificado, la proxima ejecucion de cualquiera de los metodos de lectura, inciara desde ese punto

````python
# Metodo readlines

ruta = "/content/drive/MyDrive/heroes_marvel.txt"

archivo = open(ruta,  "r")

# Vamos a omitir a Iron Man del listado
numero_de_bytes = len("Iron Man\n")
print(f"Debemos desplazarnos del inicio {numero_de_bytes} bytes para saltearnos a Iron Man de la lista")

archivo.seek(numero_de_bytes)

contenido = archivo.readlines()

print(f"Contenido del archivo: {contenido}")
print(f"Tipo de dato de contenido: {type(contenido)}")
````

Funciono! Le dijimos a seek que 9 bytes.

Si quisieras volver a recorrer el archivo, solo ponele 0 en `seek`. Eso hara que el apuntador vuelva al incio 😀

Para recuperar los datos, solo tenemos que cambiar el modo de apertura y usar una funcion distinta: `load`

Como su nombre lo indica, esta funcion va a "cargar" la informacion del archivo y va a devolver un diccionario

Aprovechando el momento, voy a mostrarte una forma mas facil de manejar la apertura y el cierre de archivos.

Python nos otorga una instruccion que nos permite abrir un archivo, guardarlo en una variable (como venimos haciendo hasta ahora) y no nos tenemos que preocupar de cerrar el archivo porque esto lo hace por nosotros.

La instruccion se llama `with`

````python
import json

with open("/content/drive/MyDrive/empleados.json", "r") as archivo:

  data = json.load(archivo)

  print(data)
  print(type(data))
````

Como podes ver, todo sigue funcionando igual 😀

# CONOCIENDO LOS RESTANTES MODOS DE APERTURA

Como hemos visto, contamos con otros modos de apertura.

Hasta ahora aprendimos que `"w"` solo sirve para escribir informacion dentro del archivo borrando su contenido previo.

`"a"` tambien solo puede escribir (misma situacion que `"w"`) pero sin borrar el contenido del archivo.

Ambos modos crear en archivo en caso de no existir.

Tambien aprendimos quie `"r"` solo sirve para leer el contenido de un archivo. Si el archivo no existe, obtenemos un error

## MODOS DE APERTURA COMBINADOS

Si quisieramos leer y escribir sin tener que abrir el archivo dos veces, tendriamos que hacer uso de alguno de los siguientes modos:

+ `w+` : permite leer y escribir en un archivo. Borra el contenido previo. Si no existe el archivo lo crea.

+ `a+` : permite leer y escribir en un archivo. No borra el contenido previo. Si no existe el archivo lo crea.

+ `r+` : permite leer y escribir en un archivo. No borra el contenido previo. Si el archivo no existe, el programa falla


Estos modos nos permiten utilizar tanto las funciones para `write` como las de `read`/`readline`/`readlines`

El unico problema es que la informacion que insertemos con el metodo `write` no se podra leer con los metodos de `read` hasta que no se abra nuevamente el archivo.


Finalmente nos queda conocer un nuevo modo de apertura: `x`

Este es un modo nuevo y especial porque viene a salvarnos de varios problemas a futuro.

El modo de apertura `x` crea y escribe contenido dentro de un archivo SI y SOLO SI, el archivo no existe en la ruta indicada

Si el archivo existe, el programa falla.

Este modo es muy utilizado cuando queremos proteger los archivos ya existentes en la maquina donde queremos ejecutar nuestro programa.
