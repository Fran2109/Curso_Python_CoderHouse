## Conjuntos

Un `set` (conjunto), es una coleccion no ordenada de objetos no repetidos.

Como otras colecciones de Python, los conjuntos son heterogeneos, es decir, que no requiere que todos los elementos sean del mismo tipo.

La unica restriccion es que los objetos que almacene un `set`, deben ser inmutables. Asi que solo son admitidos: numeros, cadenas, y tuplas

Para crear un `set` tenemos dos formas:

* Crear un set vacio
* Crear un set con elementos

```python
# set con elementos
conjunto1 = {1,2,3,4,5, "Hola mundo", (1,2,3,4)}

# set vacio
conjunto2 = set()

print(conjunto1)
```

A diferencia de las listas que usabamos [] y las tuplas que usabamos () para crearlas vacias, en los conjuntos no podemos.

Es necesario que utilizamos el tipo de dato `set()` para crearlas vacias.

### Conversion

Es posible convertir una lista o tupla a un `set` sin problemas

```python
lista = [1,2,3,4,5,5,5,5,5]

conjunto = set(lista)
print(f"Lista original: {lista}. Nuevo conjunto: {conjunto}")

tupla = (1,1,1,1,2,2,2,3,3,3)

conjunto = set(tupla)
print(f"Tupla original: {tupla}. Nuevo conjunto: {conjunto}")
```

Tambien podemos convertir un `set` en otras colecciones

```python
conjunto = {1,2,3,4,5}

lista = list(conjunto)
print(f"Nueva lista: {lista}. Tipo de dato: {type(lista)}")

tupla = tuple(conjunto)
print(f"Nueva lista: {tupla}. Tipo de dato: {type(tupla)}")
```

Seguramente algo te llamo la atencion. Todos los elementos repetidos, desaparecieron.

Como te dije antes, los conjuntos almacenan elementos unicos, es decir, no se repiten.

### List vs Set

Si bien pueden verse similares, las listas y los conjuntos son diferentes.

No hablo solo de los elementos unicos.

Al no existir un orden dentro de los conjuntos, no puedo recorrerlos por medio de indices como si lo hacia con las listas.

Por lo anterior, tampoco puedo aplicar slicing sobre sus elementos

```python
conjunto = {1,2,3,4}

print(conjunto[4])
```

## Funciones de conjuntos

Los conjuntos, como las listas y tuplas, poseen funciones integradas que permiten hacer cosas interesantes.

### ADD

Como te podras imaginar, la funcion `add()` nos permite AGREGAR elementos a nuestro conjunto.

Esta funcion recibe como argumento un solo elemento: un numero o una cadena o una tupla, y lo incorpora al conjunto

[Demo](https://giphy.com/gifs/python-add-Bzt07KxBgx4et2KVSz/fullscreen)

```python
conjunto = {1,2,3,4,"hola mundo"}

# Agregamos literales
conjunto.add(5)

print(f"Contenido del conjunto: {conjunto}")


# Agregamos variables
# Python antes de agregarla, la reemplaza por el valor que contiene el objeto
# al que esta referenciando
resultado = (2**2+4)/2+5
conjunto.add(resultado)

print(f"Contenido del conjunto: {conjunto}")

# Agregamos operaciones
# Python antes de agregar el valor, resolvera la operacion
# Y almacenara el resultado
conjunto.add(10+5**3)

print(f"Contenido del conjunto: {conjunto}")
```

### UPDATE

Esta funcion es similar a `add()` pero con la particularidad de que nos permite agregar mas de un elemento a la vez.

Como argumento recibe un objeto iterable (lista, tupla, etc.)

[Demo](https://giphy.com/gifs/update-python-eCBuXOcdU5wEusHkiK/fullscreen)

```python
conjunto = {1,2,3,4,5}

print(f"Contenido del conjunto: {conjunto}")

# Le pasamos una lista como argumento
conjunto.update([10,20,30])
print(f"Contenido del conjunto: {conjunto}")

# Le pasamos el resultado de la funcion range()
conjunto.update(range(1,10))
print(f"Contenido del conjunto: {conjunto}")

# Le pasamos el resultado de la funcion enumerate()
conjunto.update(enumerate([51,52,53]))
print(f"Contenido del conjunto: {conjunto}")

# Le pasamos una tupla
conjunto.update((61,62,63))
print(f"Contenido del conjunto: {conjunto}")
```

### LEN

`len()` no es una funcion propia de ninguna coleccion, sino mas bien es una funcion mas general de Python.

Recibe cualquier coleccion (lista, tupla, conjunto, cadena, etc) y retorna la cantidad de elementos dentro de la misma

```python
conjunto = {10,20,30,40,50}

print( len(conjunto) )
```

### DISCARD

`discard()` es una funcion que nos permite DESCARTAR (como su nombre indica) elementos de nuestro conjunto.

Recibe como argumento un solo elemento.

Si el elemento existe, lo elimina del conjunto

Si el elemento no existe, simplemente sigue con el resto del programa.


```python
conjunto = {1,2,3,4,5}

print(f"Contenido del conjunto: {conjunto}")

# Borramos un elemento que existe
conjunto.discard(3)
print(f"Contenido del conjunto: {conjunto}")

# Borramos un elemento que NO existe
conjunto.discard(10)
print(f"Contenido del conjunto: {conjunto}")
```

### REMOVE

`remove()` funciona de manera similar a `discard()`.

Recibe un solo elemento como argumento.

La diferencia es que si a `remove()` le damos un elemento que no existe dentro del conjunto, el programa falla.

```python
conjunto = {1,2,3,4,5}

print(f"Contenido del conjunto: {conjunto}")

# Borramos un elemento que existe
conjunto.remove(3)
print(f"Contenido del conjunto: {conjunto}")

# Borramos un elemento que NO existe
conjunto.remove(10)
print(f"Contenido del conjunto: {conjunto}")
```

### IN

La instruccion `in`, al igual que la funcion `len()`, no son propias de los conjuntos.

Esta instruccion es de Python.

Cuando lo usamos fuera de un `for`, esta instruccion nos permite determinar si un elemento existe dentro de una coleccion.

Super loco!

```python
elemento = 2
conjunto = {1,2,3,4,5}
print(f"El elemento {elemento} se encuentra en {conjunto}? {elemento in conjunto}")

# Para que veas que no solo funciona con conjuntos
elemento = 30
lista = [10,20,30,50]
print(f"El elemento {elemento} se encuentra en {lista}? {elemento in lista}")

elemento = 1
tupla = (2,4,6,8,10)
print(f"El elemento {elemento} se encuentra en {tupla}? {elemento in tupla}")
```

### CLEAR

Es una funcion que me permite "limpiar" o "vaciar" mi conjunto.

Si te preguntas cual es la diferencia con hacer `conjunto = set()`, mira el ejemplo

```python
conjunto = {1,2,3,4}

print(f"Contenido del conjunto: {conjunto}, Valor del ID: {id(conjunto)}")

conjunto.clear()
print(f"Contenido del conjunto: {conjunto}, Valor del ID: {id(conjunto)}")
```

```python
# Veamos que pasa si lo hacemos de otra manera

conjunto = {1,2,3,4}
print(f"Contenido del conjunto: {conjunto}, Valor del ID: {id(conjunto)}")

conjunto = set()
print(f"Contenido del conjunto: {conjunto}, Valor del ID: {id(conjunto)}")
```

Los IDs cambiaron en el segundo caso. Eso implica que Python cre otro objeto.

Eso tiene un coste en procesamiento, ya que al lenguaje le toma tiempo crear un objeto y asignarle los valores a los atributos ID, TYPE y VALUE.

### POP

Similar al `pop` que conocemos, esta funcion saca un elemento del conjunto eliminandolo, pero tambien lo devuelve

```python
conjunto = {10,20,1,2,3,4,5}

elemento_eliminado = conjunto.pop()

print(f"Se ha eliminado {elemento_eliminado} del conjunto. Contenido actual: {conjunto}")
```

## Diccionarios

Los diccionarios son colecciones no ordenadas, que almacenan elementos de la siguiente forma: `clave:valor`

A diferencia de las listas y tuplas cuyos elementos son accedidos por medio de indices (posiciones), los elementos en los diccionarios son accedidos mediante la `clave`

Estas claves pueden ser de tipo `int` o `string`, y los valores pueden ser de cualquier tipo

```python
# Crear un diccionario vacio
info_usuario = {} # tambien podemos usar dict()

# Crear un diccionario con clave valor
info_usuario = {"nombre": "Pedro", "apellido": "Laguna", "edad": 35, "email": "plaguna@email.com", 15: 123}

# Acceder a los valores del diccionario
print(f"Nombre: {info_usuario['nombre']}")
print(f"Apellido: {info_usuario['apellido']}")
print(f"Edad: {info_usuario['edad']}")
print(f"Email: {info_usuario['email']}")
print(f"Clave numerica: {info_usuario[15]}")
```

### MUTABILIDAD

Los diccionarios son mutables. Todo? No, solo los valores.

Las claves solo pueden ser creadas o borradas.

```python
precios_cryptos = {"BTC": 43600.78, "KDA": 8.8, "BORA": 1.32}
print(f"Contenido del diccionario: {precios_cryptos}")

precios_cryptos["KDA"] = 7.7
print(f"Contenido del diccionario: {precios_cryptos}")
```

```python
# Podemos usar cualquier operador de asignacion

precios_cryptos = {"BTC": 43600.78, "KDA": 8.8, "BORA": 1.32}
print(f"Contenido del diccionario: {precios_cryptos}")

# Apliquemos un recargo del 5% a todas las monedas por operar en nuestra app
precios_cryptos["BTC"] *= 1.05
precios_cryptos["KDA"] *= 1.05
precios_cryptos["BORA"] *= 1.05

print(f"Contenido del diccionario: {precios_cryptos}")
```

## FUNCIONES DE DICCIONARIOS

Los diccionarios, como otras colecciones, cuentan con funciones integradas

### ADD

No hay como tal una funcion `add` para los diccionarios.
La forma de agregar elementos es muy sencilla.

Simplemente colocamos la clave que queremos agregar entre corchetes, seguido del valor que queremos guardar

```python
precios_cryptos = {"BTC": 43600.78, "KDA": 8.8, "BORA": 1.32}
print(f"Contenido del diccionario: {precios_cryptos}")

precios_cryptos["DFI"] = 3.31
print(f"Contenido del diccionario: {precios_cryptos}")
```

Las claves deben ser unicas. Si indicamos una clave que ya existe, Python reemplazara el valor que contenia dicha clave.

### UPDATE

Este metodo permite crear/actualizar los pares clave:valor en una sola instruccion.

Puede recibir los datos de 2 formas:
1. Un diccionario
2. Una lista/tupla de tuplas donde la primer posicion de la tupla sea la clave, y la segunda el valor

```python
precios_cryptos = {"BTC": 43600.78, "KDA": 8.8, "BORA": 1.32}
print(f"Contenido del diccionario: {precios_cryptos}")

# DFI ya existe, por lo que Python reemplazara su valor
precios_cryptos.update({"DFI": 5.30, "WEMIX": 6.33})
print(f"Contenido del diccionario: {precios_cryptos}")

# Actualizando con tuplas
precios_cryptos.update([("PROM", 10.69),("ORN", 4.29)])
print(f"Contenido del diccionario: {precios_cryptos}")
```

### LEN

Como dijimos antes, `len` es una funcion que permite obtener las longitud de elementos de una coleccion.

Un diccionario es una coleccion, por ende, podemos usar esta funcion.

```python
precios_cryptos = {"BTC": 43600.78, "KDA": 8.8, "BORA": 1.32}

print( len(precios_cryptos) )
```

### DEL

`del` es una instruccion de Python que en este caso, nos permite borrar pares clave:valor de un diccionario.

Es importante que la clave existe, de lo contrario, se generara un error.

```python
precios_cryptos = {"BTC": 43600.78, "KDA": 8.8, "BORA": 1.32}
print(f"Contenido del diccionario: {precios_cryptos}")

del precios_cryptos["BORA"]
print(f"Contenido del diccionario: {precios_cryptos}")

del precios_cryptos["MONEDA_RARA_QUE_NO_EXISTE"]
print(f"Contenido del diccionario: {precios_cryptos}")
```

### IN

El operador `in` tambien puede ser usado con diccionarios y funciona similar a como funciona en las otras colecciones.

La unica diferencia es que lo que sea que comprobara si el elemento por el que estamos preguntando, existe dentro de las `claves` y no dentro de los valores.

```python
marcas = {"Ford":["Ecosport", "Mustang", "Raptor"], "Chevrolet": ["Camaro", "Onix"], "Ferrari": ["California", "F50", "F40"]}

if "Fiat" in marcas:
  print("La marca Fiat, existe en el diccionario")
else:
  print("Esa marca NO existe en el diccionario")

if "Ford" in marcas:
  print("La marca Ford, existe en el diccionario")
else:
  print("Esa marca NO existe en el diccionario")

```

### CLEAR

Tal como sucede en los conjuntos, la funcion `clear` nos permite eliminar todos los pares clave:valor sin necesidad de crear un nuevo objeto

```python
vehiculos = {"Ford":["Ecosport", "Mustang", "Raptor"], "Chevrolet": ["Camaro", "Onix"], "Ferrari": ["California", "F50", "F40"]}

print(f"Contenido del diccionario: {vehiculos}. ID: {id(vehiculos)}")

vehiculos.clear()
print(f"Contenido del diccionario: {vehiculos}. ID: {id(vehiculos)}")
```

### POP
La funcion `pop`, a diferencia del `pop` de los conjuntos, nos permite sacar un elemento del diccionario y devolverlo.

Recibe como argumento una clave.

Si la clave no existe, arrojara un error

```python
vehiculos = {"Ford":["Ecosport", "Mustang", "Raptor"], "Chevrolet": ["Camaro", "Onix"], "Ferrari": ["California", "F50", "F40"]}
print(f"Contenido del diccionario: {vehiculos}")

vehiculos_eliminados = vehiculos.pop("Ferrari")

print(f"""Contenido del diccionario: {vehiculos}.
          Eliminados: {vehiculos_eliminados}""")
```
