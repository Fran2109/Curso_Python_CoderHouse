# Cadenas

Hasta el momento no hemos conocido funciones integradas de las cadenas. Asi que es un buen momento para introducirlas

### UPPER

El metodo `upper`, es una funcion que devuelve una cadena con todos los caracteres en mayusucula

````python
# Usando variables
cadena = "hola mundo"

cadena_modificada = cadena.upper()

print(f"Cadena Original: {cadena} -- Cadena Modificada: {cadena_modificada}")

# Usando literales
cadena_modificada = 'bienvenid@s coders!'.upper()

print(f"Valor de la cadena: {cadena_modificada}")
````

### LOWER

Este metodo hace lo contrario a `upper`: devuelve una cadena con todos los caracteres en minuscula

````python
# Usando variables
cadena = "hOla MunDo"

cadena_modificada = cadena.lower()

print(f"Cadena Original: {cadena} -- Cadena Modificada: {cadena_modificada}")

# Usando literales
cadena_modificada = 'bienVEnid@s Coders!'.lower()

print(f"Valor de la cadena: {cadena_modificada}")
````

### CAPITALIZE

El caso de `capitalize` es distinto. Este metodo devuelve una cadena con el primer caracter en mayuscula, y los restantes en minuscula

````python
# Usando variables
cadena = "hOla MunDo"

cadena_modificada = cadena.capitalize()

print(f"Cadena Original: {cadena} -- Cadena Modificada: {cadena_modificada}")

# Usando literales
cadena_modificada = 'bienVEnid@s Coders!'.capitalize()

print(f"Valor de la cadena: {cadena_modificada}")
````

### TITLE

Similar a `capitalize`, `title` es un metodo que devuelve una cadena donde el primer caracter de CADA PALABRA esta en mayuscula, y los restantes caracteres en minuscula

````python
# Usando variables
cadena = "hOla MunDo"

cadena_modificada = cadena.title()

print(f"Cadena Original: {cadena} -- Cadena Modificada: {cadena_modificada}")

# Usando literales
cadena_modificada = 'bienVEnid@s Coders!'.title()

print(f"Valor de la cadena: {cadena_modificada}")
````

La funcion `find` nos permite encontrar la primer aparicion de una cadena dentro de otra.

Es importante tener en cuenta que esto devuelve la **primer aparicion** de la cadena.

### SPLIT

El metodo `split` divide la cadena original en partes, tomando como referencia una subcadena.


Si le indicamos una cadena, dividira el string en base a eso.
Si no le indicamos nada, utilizara la cadena " " (espacio) para dividir el string



````python
provincias = "Salta,Jujuy,Chaco,Formosa,Misiones"

# Dividimos el string a partir de las comas
listado_provincias = provincias.split(",")

for provincia in listado_provincias:
  print(provincia)
````

````python
# Dividimos el string a partir de los espacios
mis_calificaciones = "Matematicas 10 10 9"

listado_calificaciones = mis_calificaciones.split()

print(f"Tipo de dato de 'listado_calificaciones': {type(listado_calificaciones)}")
print(f"Nombre de la materia: {listado_calificaciones[0]}, Calificaciones obtenidas: {listado_calificaciones[1:]}")
````

### JOIN

A diferencia del `split` que divide, el metodo `join` como su nombre lo indica, UNE cadenas a partir de un string en comun

Esta funcion recibe una coleccion de elementos (lista, tupla, string) y retorna una nueva cadena.

Esta nueva cadena tiene la particularidad de que entre cada uno de los elementos, se encontrara la cadena que llama al metodo.

Mira el siguiente ejemplo para entenderlo mejor 🔽

````python
listado = ["administracion", "email.com"]

correo = "@".join(listado)

print(f"Correo generado: {correo}")



# ---------------------------
cryptos = ("BTC", "ETH", "DOGE", "ADA")

cadena_union = "+"

nueva_cadena = cadena_union.join(cryptos)

print(nueva_cadena)
````

## STRIP

`strip` es un metodo que nos permite quitar ciertos caracteres del inicio y fin de nuestra cadena, devolviendo una nueva sin estos.

Si no le indicamos nada entre parentesis, por defecto los quitara los espacios " " que se encuentren al inicio y al final.

Esta funcion es muy utilizada para limpiar nuestra cadena de caracteres no deseados.

Tengamos en cuenta que solo podemos quitar los de inicio y fin. Si los caracteres se encuentran en el medio, esta no es la funcion que vas a necesitar.

````python
cadena = "-------------Hola mundo -------------------"

cadena_nueva = cadena.strip("-")

print(cadena_nueva)
````

# LISTAS

Hemos conocido algunos metodos de las listas un poco basicos para poder trabajar, pero no creas que son todos.

### CLEAR

Anteriormente vimos una forma de vaciar el contenido de las listas:

```python
lista  = [1,2,3,4]

lista = []
```

Si bien esta forma es rapida, tiene un costo alto para Python. 

Mira los id's en el siguiente codigo:

````python
lista  = [1,2,3,4]

print(f"Lista con elementos: {id(lista)}")

lista = []

print(f"Lista vacia: {id(lista)}")
````

El id cambio. 

Esto implica que Python tuvo que crear un nuevo objeto y asociarlo a la variable.

Si queres saber que tiene que hacer Python para crear un objeto, te invito a leer el siguiente Colabs

Para solucionar esto, las listas incorporaron un metodo que elimina todos los elementos dentro de ellas sin necesidad de crear nuevos objetos:


````python
lista  = [1,2,3,4]

print(f"Lista con elementos: {id(lista)}")

lista.clear()

print(f"Lista vacia: {id(lista)}")
````

### EXTEND

Un caso similar se da cuando queremos agregar varios elementos a nuestra lista de la siguiente forma:

```python
lista  = [1,2,3,4]

lista = lista + [10,20,30]

print(f"Lista vacia: {id(lista)}")

```

Esto hace que python tenga que crear un objeto que contenga los elmentos de ambas listas.


````python
lista  = [1,2,3,4]

print(f"Lista original: {id(lista)}")


lista = lista + [10,20,30]

print(f"Lista con mas elementos: {id(lista)}")
````

Asi que Python incorporo la funcion `extend` que permite EXTENDER la lista agregando mas elementos, sin necesidad de crear un nuevo objeto

````python
lista  = [1,2,3,4]

print(f"Lista original: {id(lista)}")


lista.extend([10,20,30])

print(f"Lista con mas elementos: {id(lista)}")
````

### INSERT

Te acordas de `append`? Esta funcion nos permitia agregar elementos a nuestra lista. Pero habia un problema: los elementos iban al final.

Y si queria agregarlos en el medio? Tenia que hacer un slicing y una concatenacion. El problema es que eso es ineficiente


````python
lista  = [1,2,3,4]

print(f"Lista original: {id(lista)}")

lista = lista[:2] + [10] + lista[2:]

print(f"Lista con mas elementos: {id(lista)}")
print(f"Lista: {lista}")
````

El metodo `insert` viene a resolver eso.

Solo necesitamos indicarle dos cosas: el indice donde queremos el elemento y el elemento a agregar



````python
lista  = [1,2,3,4]

print(f"Lista original: {id(lista)}")

lista.insert(2,10)

print(f"Lista con mas elementos: {id(lista)}")
print(f"Lista: {lista}")
````

### REVERSE

El metodo `reverse` reordena la lista de elementos. De que forma? Invirtiendola

````python
lista  = [1,2,3,4]

print(f"Lista original: {lista}")

lista.reverse()

print(f"Lista original: {lista}")
````

### SORT

Alguna vez quisiste reordenar los elementos? `sort` es la funcion que necesitas.

Llamando a este metodo, los elementos de la lista se ordenaran de menor a mayor.

Si queres ordenarlos de mayor a menor, hay que pasarle un valor extra.

````python
# Ordenando de MENOR a MAYOR

lista  = [10,50,1,2,5,-10,-100]

print(f"Lista original: {lista}")

lista.sort()

print(f"Lista ordenada: {lista}")
````

````python
# Ordenando de MAYOR a MENOR

lista  = [10,50,1,-10,2,5,-100]

print(f"Lista original: {lista}")

# Reverse es una variable que se encuentra dentro de la funcion sort()
# Si le pasamos el valor True, podemos hacer que ordene la lista y luego la invierta por nosotros
lista.sort(reverse=True)

print(f"Lista ordenada: {lista}")
````

````python
# Otra forma de hacerlo

# Ordenando de MAYOR a MENOR

lista  = [10,50,1,2,5,-10,-100]

print(f"Lista original: {lista}")

# Reverse es una variable que se encuentra dentro de la funcion sort()
# Si le pasamos el valor True, podemos hacer que ordene la lista y luego la invierta por nosotros
lista.sort()
lista.reverse()

print(f"Lista ordenada: {lista}")
````

### COPY

Supongamos que queremos copiar un conjunto en otra variable porque queremos modificar el contenido de este pero no queremos perder los datos.

Asi que se nos ocurre la idea de hacer esto

````python
set1 = {1,2,3,4,5}

set2 = set1

print(f"Valor del set1 {set1}")
print(f"Valor del set2 {set2}")
````

````python
# Ahora modificamos el set1 como teniamos pensado hacer

set1.update([10,20,30])

# Vamos a imprimir ambos conjuntos ahora
print(f"Valor del set1 {set1}")
print(f"Valor del set2 {set2}")
````

````python
# Oh no!! Se modifico el set2.. pero por que?

# Miremos de cerca los id's de ambos conjuntos
print(f"Valor del ID de set1 {id(set1)}")
print(f"Valor del ID de set2 {id(set2)}")
````

Son identicos.

Esto se debe a que Python, para optimizar el uso de memoria, hace que ambas variables apunten al mimso objeto.

Y esto no solo pasa con los conjuntos, sino que tambien pasa con las listas, y los diccionarios.

Esto es un problema enorme!!

Pero tranquilo/a! Tenemos una solucion bajo la manga: `copy`

````python

# Iniciamos de la misma forma
set1 = {1,2,3,4,5}

# Indicamos por medio del metodo que queremos una copia de los elementos
# Esto devolvera una coleccion del tipo `set` y se lo asignara a set2
set2 = set1.copy()

print(f"Valor del set1 {set1}")
print(f"Valor del set2 {set2}")


# Ahora modificamos el set1 como teniamos pensado hacer
set1.update([10,20,30])

# Vamos a imprimir ambos conjuntos ahora
print(f"Valor del set1 {set1}")
print(f"Valor del set2 {set2}")

# Miremos de cerca los id's de ambos conjuntos tambien
print(f"Valor del ID de set1 {id(set1)}")
print(f"Valor del ID de set2 {id(set2)}")
````

Ahora efectivamente las variables apuntan a objetos distintos y ya no tenemos que preocuparnos por la perdida de los valores a la hora de modificar set1 ya que mantenemos una copia en set2 😀

Este metodo tambien puede ser utilizado en las demas colecciones


````python
# Ejemplo con listas

lista1 = [1,2,3]
lista2 = lista1

# Vamos a imprimir ambas listas
print(f"Valor del lista1 {lista1}")
print(f"Valor del lista2 {lista2}")

# Miremos de cerca los id's de ambas listas
print(f"Valor del ID de lista1 {id(lista1)}")
print(f"Valor del ID de lista2 {id(lista2)}")


# Si usamos el .copy()
lista2 = lista1.copy()

# Vamos a imprimir ambas listas
print(f"Valor del lista1 {lista1}")
print(f"Valor del lista2 {lista2}")

# Miremos de cerca los id's de ambas listas
print(f"Valor del ID de lista1 {id(lista1)}")
print(f"Valor del ID de lista2 {id(lista2)}")
````

### ISDISJOINT

Esto va de la mano con la teoria matematica de conjuntos.


Supongamos que contamos con lo siguiente:

```python
A = {7,1,5,3}
B = {2,4,5,3}
```

image.png

La teoria dice que dos conjuntos son DISJUNTOS cuando su interseccion es VACIA, es decir, NO COMPARTEN ELEMENTOS EN COMUN.

Interesante!

La funcion `isdisjoint` establece eso, devolviendo un valor booleano.

Esta funcion espera recibir una coleccion. Puede ser un conjunto, tupla o lista.

Si no tienen elementos en comun, es decir, ningun elemento de A esta en B y viceversa, la funcion devolvera True.

En el caso que compartan al menos un valor, la funcion devolvera False

````python
A = {7,1,5,3}
B = {2,4,5,3}

print(f"Los conjuntos A y B son disjuntos? { A.isdisjoint(B) }")
print(f"Los conjuntos A y B son disjuntos? { B.isdisjoint(A) }")


# Podemos pasarle otras colecciones


# Lista
B = [4,8,10]
print(f"Los conjuntos A y B son disjuntos? { A.isdisjoint(B) }")


# Tupla
B = (4,8,10)
print(f"Los conjuntos A y B son disjuntos? { A.isdisjoint(B) }")

````

````python
# Cuidado: este metodo solo esta disponible para los conjuntos.

A = [7,1,5,3]
B = {2,4,5,3}

print(f"Los conjuntos A y B son disjuntos? { A.isdisjoint(B) }")
````

### ISSUBSET

Supongamos el siguiente caso:

```python
A = {1,2,3}
B = {1,2,3,4,5}
```

image.png

La teoria dice: Dados dos conjuntos, A y B, decimos que el conjunto A es un subconjunto de B SI Y SOLO SI todos los elementos de A se encuentran en B


Para determinar esto, contamos con la funcion `issubset` que al igual que la anterior, recibe una coleccion y devuelve True en caso de que la coleccion sea un subconjunto, y False en caso de no serlo.

Aca si importa el orden en que ejecutemos la operacion.

No es lo mismo preguntar si A es subconjunto de B, que preguntar si B es subconjunto de A.

Cuidado!

````python
A = {1,2,3}
B = {1,2,3,4,5}


print(f"El conjunto B es un subconjunto de A? { B.issubset(A) }")
print(f"El conjunto A es un subconjunto de B? { A.issubset(B) }")
````

### ISSUPERSET

Similar a lo anterior, establece si un conjunto contiene a otro.

PD: en el anterior preguntabamos si el conjunto A estaba contenido dentro de B

Ahora vamos a preguntar si B contiene a A


Es una operacion media parecida

````python
A = {1,2,3}
B = {1,2,3,4,5}


print(f"El conjunto B contiene al conjunto A? { B.issuperset(A) }")
print(f"El conjunto A contiene al conjunto B? { A.issuperset(B) }")
````

### UNION


La union en matematicas, conciste en un conjunto que reune todos los valores de A y B, siendo A y B dos conjuntos.

No hace falta que tengan elementos en comun. Pueden ser distintos.

Esta funcion recibe una coleccion de elementos y retorna un conjunto con todos los elementos que estaban en el conjunto previamente + los elementos de la coleccion indicada.

El conjunto original no se modifica!

````python
A = {1,2,3}
B = [4,5,6]


nuevo_conjunto = A.union(B)

# Como vemos, A no se modifica
print(f"Conjunto original: {A}. Nuevo conjunto: {nuevo_conjunto}")
````

### DIFFERENCE

Dados dos conjuntos A y B, la diferencia del conjunto A respecto a B, es un nuevo conjunto que contiene todos los elementos que estan en A pero no en B.

Depende quien llame al metodo, el resultado varia.

````python
A = {1,2,3}
B = {3,4,5}

C = A.difference(B)
print(f"Los elementos que estan en A pero no en B son: {C}")

C = B.difference(A)
print(f"Los elementos que estan en B pero no en A son: {C}")
````

Como vemos, en el primer caso A solo tiene 1 y 2 como elementos unicos

### DIFFERENCE_UPDATE

Esta funcion hace lo mismo que la anterior pero ademas, modifica el contenido del conjunto (de ahi la parte de **update**)

Con que valores lo actualiza? Con lo que sea que devuelva `difference`

Otra cosa a tener en cuenta es que no devuelve un conjunto

````python
A = {1,2,3}
B = {3,4,5}

print(f"Valor de A: {A}, Valor de B: {B}")
print("Calculando diferencia y aplicando cambios en A")
A.difference_update(B)
print(f"Valor de A: {A}, Valor de B: {B}")

# Esto es solamente para dividir los mensajes que se imprimen. Para que no salgan
# muy pegados y sea mas facil de leer
print("="*10) 

A = {1,2,3}
B = {3,4,5}

print(f"Valor de A: {A}, Valor de B: {B}")
print("Calculando diferencia y aplicando cambios en B")
B.difference_update(A)
print(f"Valor de A: {A}, Valor de B: {B}")
````

### INTERSECTION

Dados dos conjuntos A y B, la interseccion se define como el conjunto de elementos que comparten tanto A como B, es decir que se encuentran en AMBOS.

Este metodo devuelve un nuevo conjunto y NO modifica el original

````python
A = {1,2,3}
B = {3,4,5}


C1 = A.intersection(B)
C2 = B.intersection(A)

# En este caso, no importan quien llame al metodo (siempre y cuando sea un conjunto)
# El resultado es el mismo
print(f"A interseccion B: {C1}")
print(f"B interseccion A: {C2}")
````

### INTERSECTION_UPDATE

Similar a lo que pasa con `difference_update`, este metodo calcula la interseccion y cambia el contenido del conjunto que la llamo.

````python
A = {1,2,3}
B = {3,4,5}

print(f"Valor de A: {A}")
A.intersection_update(B)
print(f"Valor de A: {A}")
````

### GET

Como recordaras, si intentabamos acceder a un elemento dentro de un diccionario mediante una clave que no existia, el programa fallaba 🔽

````python
diccionario = {"Marca": "Ford", "Modelo": "Focus", "Kms": 110000, "UnicoDuenio": True}

# Clave que no existe
print(diccionario["Motor"])
````

Alguien penso en esto y lo resolvio con un metodo sencillo: `get`

Este metodo recibe dos cosas: la clave, y un valor

Este metodo intentara acceder al elemento mediante la clave, pero en caso de no existir, nos devolvera el valor que hayamos indicado como segundo argumento a la funcion

````python
diccionario = {"Marca": "Ford", "Modelo": "Focus", "Kms": 110000, "UnicoDuenio": True}

# Clave que no existe
print(diccionario.get("Motor", 2.0))


diccionario = {"Marca": "Ford", "Modelo": "Focus", "Kms": 110000, "UnicoDuenio": True, "Motor": 1.6}

# Clave que AHORA si existe
print(diccionario.get("Motor", 2.0))
````

### KEYS

Si por alguna razon necesitaramos un listado de todas las claves de un diccionario, contamos con el metodo `keys`

Vemos un ejemplo de donde utilizarlo:

````python
diccionario = {"Marca": "Ford", "Modelo": "Focus", "Kms": 110000, "UnicoDuenio": True}
listado_claves = diccionario.keys()

# Menu de opciones
print("+"*5, "MENU", "+"*5)

for opcion in listado_claves:
  print(f">> {opcion}")

opcion = input("Introduzca una opcion: ")

valor = diccionario.get(opcion, "????")

print(f"Valor obtenido: {valor}")
````

### VALUES

Similar a `keys`, este metodo nos devuelve un listado con todos los valores dentro de nuestro diccionario.

````python
diccionario = {"Marca": "Ford", "Modelo": "Focus", "Kms": 110000, "UnicoDuenio": True}
listado_valores = diccionario.values()

print(listado_valores)
````

Veamos un uso

````python
lista_vehiculos = [
                   {"Marca": "Ford", "Modelo": "Focus", "Kms": 110000, "UnicoDuenio": True},
                   {"Marca": "Ford", "Modelo": "Fiesta", "Kms": 153500, "UnicoDuenio": False},
                   {"Marca": "Ford", "Modelo": "Mondeo", "Kms": 123000, "UnicoDuenio": False},
                   {"Marca": "Fiat", "Modelo": "Uno", "Kms": 235000, "UnicoDuenio": True}
                   ]

print("========== SUPER OPORTUNIDAD ==========")
print("🚗 Nuestros mejores autos usados al mejor precio 🚗 \n")

for vehiculo in lista_vehiculos:
  # Esto lo vimos en el archivo de conjuntos
  marca, modelo, kms, unico_duenio = vehiculo.values()

  if unico_duenio:
    print(f"🚘{marca} {modelo} - KMs: {kms} - Es unico dueño? Si")
  else:
    print(f"🚘{marca} {modelo} - KMs: {kms} - Es unico dueño? No")
````

### ITEMS

Este metodo es la fusion de los dos anteriores.

Nos devuelve una lista de tuplas, donde cada tupla contiene en el indice CERO la clave y en el indice UNO el valor

Muy utilizado en un `for`

````python
vehiculo_info = {"Marca": "Ford", "Modelo": "Focus", "Kms": 110000, "UnicoDuenio": True}

print(vehiculo_info.items())
````

````python
vehiculo_info = {"Marca": "Ford", "Modelo": "Focus", "Kms": 110000, "UnicoDuenio": True}
                   

print("========== SUPER OPORTUNIDAD ==========")
print("🚗 El auto mas buscado al mejor precio 🚗 \n")


# Cada elemento de la lista es una tupla con dos elementos, asi que podemos dividirla
# y asignarle cada elemento a las variables

for clave, valor in vehiculo_info.items():
  print(f"{clave}: {valor}")
````
