## El proceso de repetir cosas

En nuestro dia a dia, nos cansamos de repetir cosas. Y si bien es cierto que no siempre lo hacemos de la misma forma, la idea sirve.

Pensa en la ruta que seguis para ir al trabajo, la rutina que ejecutas cada maana, etc.

Todas ellas, son tareas repetitivas, porque cumplen un ciclo (inicio a fin) y vuelven a comenzar.

Las computadoras vinieron a ayudar en esas tareas que requieren repetirse una y otra vez, y deben hacer siempre de la misma forma. Porque si, son buenas haciendolo.


## WHILE

La sentencia iterativa `while` (traduccion: mientras), es una instruccion que nos permite repetir un bloque de codigo (como pasaba con `if`) siempre y cuando la condicion logica que evalua, sea Verdadera (True).

Al momento en que el resultado de la condicion cambie a Falsa (False), el bloque de codigo dejara de repetirse y se seguira con el resto del programa.

Esta setencia es muy utilizada cuando desconocemos la cantidad de veces que debemos repetir un bloque de codigo. 

Veamos este ejemplo:

````python
usuario = input("Ingrese su usuario: ")

# Ahora que ya sabemos usar condiciones como PROs, podemos hacer esto
while not usuario:
  print("Debe ingresar un nombre de usuario")
  usuario = input("Ingrese su usuario: ")

print("Bienvenid@", usuario)
````

Este pequeo programa funciona asi:

1. Python solicita al usuario que ingrese algo por consola
2. La instruccion WHILE evalua la condicion. En este caso, si el usuario ingreso un string cualquiera que no sea una cadena vacia, el bloque no se ejecutara. Sin embargo, si presiono Enter solamente, la condicion dara verdadera y por lo tanto, ejecutara el codigo.
3. Suponiendo que el usuario no ingreso nada la primera vez, se imprime un mensaje de advertencia y se solicita nuevamente el dato
4. Al terminar con la ultima instruccion, volvemos al paso 2
5. Si la instruccion da False, es decir que el usuario ingresa una cadena valida, el bucle se termina y se sigue con el resto del programa.


Que pasa si nunca cambia el valor de la condicion? Simplemente entrariamos en un bucle infinito y la unica solucion es finalizar todo el programa y perder todos los datos que hayamos cargado hasta el momento.

````python
usuario = input("Ingrese su usuario: ")

# Ahora que ya sabemos usar condiciones como PROs, podemos hacer esto
while not usuario:
  print("Debe ingresar un nombre de usuario")
  Usuario = input("Ingrese su usuario: ")

print("Bienvenid@", usuario)
````

Observemos en el caso anterior como solo equivocandonos en el nombre de la variable, podemos hacer que todo nuestro programa caiga en un bucle infinito, asi que cuidado con esto.

## WHILE-ELSE

Se trata de la misma sentencia que vimos recientemente pero con un agregado: un else

_Un else? Eso no era del if?_

El `else` no solo se encuentra presente en el `if`. Hay otras sentencias que lo han adoptado para su uso. While es una

Sigue el mismo principio: solo ejecutara su bloque de codigo si la condicion evaluada diera False.

Veamos este ejemplo

````python
heroes = ["Iron Man", "Spiderman", "Wolverine", "Captain America"]

heroe_buscar = input("Ingrese su heroe: ")

# Inidice usado para recorrer la lista de heroes
indice = 0

while indice < len(heroes):

  if heroe_buscar == heroes[indice]:
    print("Heroe encontrado")

  indice += 1
  print("Seguimos buscando...")
else:
  print("Heroe no encontrado")
````

Excelente! Nuestro heroe fue encontrado! Pero....

Por que se imprimio el mensaje de "Heroe no encontrado"?

La razon es que el `else` se va a ejecutar siempre que la condicion de False.

Podemos arreglarlo? Por supuesto que si! La solucion esta en la siguiente parte 

## Instrucciones

Es probable que queramos modificar el comportamiento de nuestro bucle segun ciertas condiciones.

Esto permitiria incluso hacerlos mas eficientes.

Python nos ofrece tres instrucciones que podemos usar en bucles (y en otras partes): `break`, `continue` y `pass`



### BREAK

Como su nombre lo indica (romper), esta instruccion rompe por completo la iteracion

````python
heroes = ["Iron Man", "Spiderman", "Wolverine", "Captain America"]

heroe_buscar = input("Ingrese su heroe: ")

# Inidice usado para recorrer la lista de heroes
indice = 0

while indice < len(heroes):

  if heroe_buscar == heroes[indice]:
    print("Heroe encontrado")
    break
    
  indice += 1
  print("Seguimos buscando...")
else:
  print("Heroe no encontrado")
````

Aca pasaron 2 cosas interesantes:

1. Cuando la condicion del `if` dio Verdadera, se imprimio el mensaje y se finalizo por completo el bucle. No ejecuto ninguna de las instrucciones que vienen despues.

2. No se ejecuto la instruccion dentro del `else`

La situacion 1 se debe a la naturaleza de `break` asi que es un comportamiento esperado y DESEADO.

La situacion 2 por otra parte se debe a lo que escribi antes: Solo se va a ejecutar cuando la condicion del WHILE haya dado False. Y esto nunca paso. Rompimos el bucle con una intruccion, no con la condicion.

Ahora nuestro programa de busqueda de heroes funciona mejor porque podemos buscarlos y recorrer la lista hasta donde se encuentren y no malgastar recursos en seguir buscando cuando ya tengo lo que quiero.



### CONTINUE

Esta instruccion, a diferencia de `break`, me permite omitir las instrucciones que vengan despues de el.

Veamos

````python
calificaciones = [9,8,7,2,5,4,6,8,1,5,10]

indice = 0

while indice < len(calificaciones):
  
  if calificaciones[indice] < 7:
    indice += 1
    continue
  
  print(f"Calificacion alta: {calificaciones[indice]}")
  indice += 1
````

El bucle no se rompe, sino que se omiten todas las instrucciones posteriores a `continue`

### PASS

Esta instruccion literalmente no hace nada.

Para que quiero una instruccion que no haga nada?

A mi y a muchos programadores nos permite poner a funcionar codigo que aun no esta terminado.

Por ejemplo

````python
calificaciones = [9,8,7,2,5,4,6,8,1,5,10]

indice = 0

while indice < len(calificaciones):
  
  if calificaciones[indice] < 7:
    # Tengo que terminar esta sentencia
    pass
  else:
    print(f"Calificacion alta: {calificaciones[indice]}")
  
  print("Siguiente...")
  indice += 1
````

Sin el `pass` me veria obligado a escribir las instrucciones, o simplemente borrar toda la sentencia y rearmar mi codigo

## FOR

Esta sentencia nos permite recorrer objetos iterables como listas, tuplas, etc. y ejecutar un bloque de codigo en cada iteracion.

Esto es un `for`

```python
for valor in objeto_iterable
```

* `for` es la instruccion
* `valor` es una variable que almacenara cada elemento del objeto_iterable
* `in` es una instruccion de Python usada en este caso para asignar dentro de `valor` un elemento de `objeto_iterable`
* `objeto_iterable` es la lista, tupla, etc. que recorreremos



````python
personajes_marvel = ["Daredevil", "Black Widow", "Valkyrie", "Iron Lad", "Punisher"]

for personaje in personajes_marvel:
  print(f"El personaje es: {personaje}")
````

````python
# Podemos hacer operaciones con la variable creada en el FOR

precios = [5.95, 76.23, 22.48, 98.51, 35.92]

for precio in precios:
  # Recargo del 5%
  precio *= 1.05
  precio = round(precio, 2)
  print(precio)

print(precios)
````

En ambos casos `personaje` y `precio`, son variables que contiene una copia del valor del arreglo

La variable `personaje` puede seguir siendo usada una vez que el for finaliza. El valor que contiene es el ultimo obtenido en el for.


````python
personajes_marvel = ["Daredevil", "Black Widow", "Valkyrie", "Iron Lad", "Punisher"]

for personaje in personajes_marvel:
  print(f"El personaje es: {personaje}")

print(f"El ultimo valor obtenido fue: {personaje}")
````

El `for` va tomando de a un valor, y se lo asigna a la variable creada en el.

Termina cuando el objeto iterable se haya quedado sin elementos

### Enumerate

Se trata de una funcion de Python que recibe como argumento un objeto iterable (lista, tupla, etc) y devuelve una lista de tuplas.

Estas tuplas contienen 2 valores:

1. El primer valor corresponde al indice del elemento
2. El segundo valor corresponde al elemento en si

````python
personajes_marvel = ["Daredevil", "Black Widow", "Valkyrie", "Iron Lad", "Punisher"]

for personaje in enumerate(personajes_marvel):
  print(f"El personaje es: {personaje[1]} y se encuentra en el indice {personaje[0]}")
````

Una tecnica comun con enumerate es desestructurar las tuplas.

Sabemos que cada iteracion devolvera una tupla con dos elementos. Siempre es asi.

Asi que con Python podemos implementar esto:


````python
personajes_marvel = ["Daredevil", "Black Widow", "Valkyrie", "Iron Lad", "Punisher"]

for indice, personaje in enumerate(personajes_marvel):
  print(f"El personaje es: {personaje} y se encuentra en el indice {indice}")
````

Al declarar varias variables separadas por coma en el FOR, le indicamos a Python que "desarme" la coleccion que viene despues de in y asigne los elementos en las variables segun corresponda.

Es importante tener en cuenta el orden, ya que si nos equivocamos podriamos tener resultados no esperados

````python
personajes_marvel = ["Daredevil", "Black Widow", "Valkyrie", "Iron Lad", "Punisher"]

for personaje, indice in enumerate(personajes_marvel):
  print(f"El personaje es: {personaje} y se encuentra en el indice {indice}")
````

## Range

Que pasa en los escenarios donde necesito ejecutar un bloque de codigo N veces?

Python nos provee una funcion para "simular" estos casos. Se trata de `range` cuya tarea es la de generar una coleccion de numeros enteros inmutables.

Tenemos 3 formas de usar range()

````python
# Forma 1

# Si le pasamos un solo valor a esta funcion, generara una tupla
# con numerosa que van del CERO hasta el numero anterior al indicado
for indice in range(10):
  print(f"Valor del indice {indice}")
````

````python
# Forma 2

# Si le indicamos dos valores a la funcion
# Tomara el primero de ellos como el INICIO y el segundo como el FIN
# Recordemos que el FIN es el numero anterior al indicado

for indice in range(2, 10):
  print(f"Valor del indice {indice}")
````

````python
# Forma 3

# Si le indicamos tres valores, tomara el primero como INICIO
# el segundo como FIN y el tercero como el "PASO" (cantidad de numeros que saltamos)

for indice in range(0, 10, 2):
  print(f"Valor del indice {indice}")
````

Esta funcion es mas poderosa de lo que parece ya que no crea una lista o tupla como tal, sino que cada vez que el `for` le pide mas digitos, este lo genera sumandole el *paso* al valor actual y lo devuelve.

Esta operacion lo hace hasta que el valor actual alcanza o supera el valor de FIN.

Una lista de 10000 elementos necesita de mucha memoria antes de poder usarla. `range` solo necesita un espacio de memoria relativamente pequeo para almacena 1 solo numero.

## FOR-ELSE

Al igual que en `while`, el `for` admite el `else` para ejecutar un bloque de codigo.

En este caso no evaluamos condiciones, asi que el `else` se ejecutara solo cuando el `for` haya finalizado porque el objeto iterable se quedo sin elementos.


````python
acciones = ["AAPL", "ADBE", "AMZN", "BBV", "CAT", "GRMN"]

accion_buscar = input("Ingrese el nombre de la accion (p.e. AAPL): ")

for accion in acciones:
  if accion == accion_buscar:
    print("Accion encontrada")
    break
  print("Buscando...")
else:
  print("No tenemos registro de esa accion")
````

Tanto dentro de un `for` como en `while` podemos declarar cualquier instruccion que hayamos visto hasta ahora.

Podemos tener un `for` dentro de otro, un `while` con un `for` anidado, if-elif-else dentro de un `for` o `while`.

Podemos inclusive usar `continue`, `pass` y `break` dentro de un `for`.
