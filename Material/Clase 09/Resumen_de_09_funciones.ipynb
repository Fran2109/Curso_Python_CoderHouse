## CREAR FUNCIONES

Para hacer esto, Python nos da una instruccion: `def` (definition)

Esta funcion nos sirve para crear nuestras propias funciones que se adapten al problema que queremos resolver.

Crear las funciones es sencillo

```python
def nombre_de_la_funcion(parametro1, parametro2):
  # Aca defino mi algoritmo
  return valor
```

* Debemos de darle un nombre a la funcion
* Debemos de indicarle que valores va a recibir y como se van a llamar dentro de nuestro codigo
* Debemos de declarar todo el codigo que queremos que se ejecute cuando nuestra funcion sea usada
* Debemos de indicar despues de la instruccion `return` que va a devolver mi funcion


Veamos un ejemplo

````python
# <---  Presiona primero este boton antes del siguiente cuadro de codigo

def saludar():
  print("Hola coder!!")
````

````python
saludar()
````

Si queremos que reciba datos:

````python
def saludar_coder(nombre):
  print(f"Hola {nombre.title()}! Espero que estes disfrutando el curso")
````

````python
nombre_persona = input("Ingrese su nombre: ")
saludar_coder(nombre_persona)
````

El nombre que se encuentra entre los parametros de la funcion (donde esta el `def`), es una variable que va a tomar el valor que le estemos pasando a la funcion.

Dentro de nuestra funcion, esa variable va a poder ser usada. Pero ojo! `nombre` solo existe dentro de la funcion

Veamos un ejemplo donde devolvamos una cadena


````python
def generador_de_correo(nombre, apellido, empresa):
  correo = nombre[0].upper() + apellido.capitalize() + "@" + empresa.lower() + ".com"
  return correo
````

````python
nombre = input("Ingrese su nombre: ")
apellido = input("Ingrese su apellido: ")
empresa = input("Ingrese su empresa: ")

correo_empleado = generador_de_correo(nombre, apellido, empresa)

print(f"Felicidades {nombre.capitalize()}, tu nuevo correo es: {correo_empleado}")
````

Esto es similar a lo que pasaba con `len()`: le pasabamos una coleccion y el devolvia un numero

Probemos hacer nuestro propio `len()`

````python
def len_coder(coleccion):
  # Inicializamos el contador en cero
  contador = 0

  # Incrementamos en 1 el contador por cada item dentro de la coleccion
  for item in coleccion:
    contador += 1
  
  # Devolvemos el valor
  return contador
````

````python
print("LEN de Python")

print(f"Lista vacia []: { len([]) }")
print(f"Cadena vacia '': { len('') }")
print(f"Tupla vacia (): { len( tuple() )} ")

print(f"Lista con elementos [1,2,3]: {len([1,2,3])}")
print(f"Cadena con elementos 'hola': {len('hola')}")
print(f"Tupla con elementos (1,2,3): {len((1,2,3))}")

print("="*20)
print("LEN de Coder")

print(f"Lista vacia []: { len_coder([]) }")
print(f"Cadena vacia '': { len_coder('') }")
print(f"Tupla vacia (): { len_coder( tuple() )} ")

print(f"Lista con elementos [1,2,3]: {len_coder([1,2,3])}")
print(f"Cadena con elementos 'hola': {len_coder('hola')}")
print(f"Tupla con elementos (1,2,3): {len_coder((1,2,3))}")
````

## Variables en Funciones

Como dijimos antes, las variables creadas dentro de la funcion, solo existen dentro de estas, por lo que el siguiente codigo daria un error

````python
def funcion_matematica(x):
  resultado = 2*x + 3
  print(resultado)

# Fuera de la funcion
print(resultado)
````

Todo lo que este pegado al margen (sin espacios a la izquierda) es de caracter global.

Esto quiere decir que puede ser usado en cualquier parte de nuestro codigo. Obviamente despues de haber sido declarado

````python
heroe = "Iron Man"

def print_heroe():
  print(f"Tu heroe es: {heroe}")
print_heroe()
````

Que pasa si una variable global se llama igual que una que existe dentro de una funcion?

````python
heroe = "Iron Man"

def print_heroe(heroe):
  heroe = heroe.title()
  
  print(f"Tu heroe es: {heroe}")

print_heroe("capitan america")
````

Python le da prioridad a la variable que se encuentra dentro de la funcion.

Si quiero usar una variable, Python primero verificara que exista dentro de la funcion, y si no existe, comienza a buscarla por los niveles de identacion superiores hasta encontrarla.

Si no la encuenta, ya sabemos que pasa

Si colocamos un `return` sin nada a su derecha, tambien obtenemos ese resultado

````python
def test(nombre):
  print(f"Hola {nombre}!")
  return

print( test("admin") )
````

Ahora te estaras preguntando _si python ya devuelve None sin que le indique el return, para que lo usaria asi solo?_

El `return` funciona casi como funciona un `break`.

Todo lo que se encuentre por debajo de un `return`, no sera ejecutado

````python
def test(nombre):
  print(f"Hola {nombre}!")
  return
  print("Pongo otro print que jamas se va a ejecutar")

print( test("admin") )
````

Este dato es muy interesante porque me permite cortar con el flujo de la funcion cuando yo quiera

````python
def buscar_en_coleccion(coleccion, dato):

  for indice, elemento in enumerate(coleccion):
    if elemento == dato:
      print(f"Elemento encontrado en el indice {indice}")
      return
  # Si termin el for, es porque jamas entro en el if
  # Por ende quiere decir que el elemento no existe en la coleccion
  # Asi que imprimimos el mensaje
  print(f"Elemento no encontrado")


lista = [1,2,3,4,5]
buscar_en_coleccion(lista, 3)
buscar_en_coleccion(lista, 10)
````

Queres que te diga algo mas interesante aun del `return`? Podemos devolver mas de un valor

````python
def buscar_en_coleccion(coleccion, dato):

  for indice, elemento in enumerate(coleccion):
    if elemento == dato:
      return True, indice, elemento

  return False, -1, None

lista = [1,2,3,4,5]
resultado = buscar_en_coleccion(lista, 4)

print(f"Resultado: {resultado}")
print(f"Tipo de dato: {type(resultado)}")
````

````python
# Recorda que podemos desempaquetar los valores. Sabemos que son 3 elementos los que devuelve
# siempre asi que no hay problema

# Si uno de los return devolviera algo distinto, si tendriamos un problema

def buscar_en_coleccion(coleccion, dato):

  for indice, elemento in enumerate(coleccion):
    if elemento == dato:
      return True, indice, elemento

  return False, -1, None

lista = [1,2,3,4,5]
encontrado, indice, elemento = buscar_en_coleccion(lista, 4)

if encontrado:
  print(f"Indice: {indice}, Elemento: {elemento}")
else:
  print("Elemento no encontrado")
````

## Momentos de una funcion


Esto es la **definicion** de una funcion
```python
def saludar():
  print("hola")
```

Esto es la **llamada** de una funcion

```python
saludar()
```

---
En la definicion de la funcion, las variables de denominan **parametros**

```python
def suma(a,b):
  return a+b
```

En la llamada de la funcion, los valores se denominan **argumentos**

```python
resultado = suma(12,2)
```

Python asignara los valores de los argumentos, segun hayamos definido los parametros

```python
def division(a,b):
  print(a/b)
```

Esta funcion dara un resultado distinto segun el orden en que le pasemos los valores.

No es lo mismo $1/3$ que $3/1$

````python
def division(a,b):
  print(a/b)

division(1,3)
division(3,1)
````

Si queremos indicar como deben asignarse los valores, podemos hacer lo siguinte (igual que en la funcion `sort`)

````python
def division(a,b):
  print(a/b)

division(b=1,a=3)
division(a=3,b=1)
````

Ahora no importa el orden en que le pasemos los valores porque se lo estamos indicando
