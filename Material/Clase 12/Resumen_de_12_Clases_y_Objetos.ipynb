## Programacion Orientada a Objetos

Para entender que es la **Programacion Orientada a Objetos** (P.O.O.) debemos de comprender primero que es un paradigma de programacion.

### Paradigma de programacion

Existen muchas interpretaciones y formas de explicar que es un paradigma.
Un paradigma de programacion no es mas que una forma de resolver un problema.

Existen varios paradigmas y cada uno tiene su propio enfoque y sus diferencias respecto a los otros. Existen dos grandes grupos: Paradigmas Declarativos y Paradigmas Imperativos



**Programacion imperativa**: 

Este paradigma agrupa a otros paradigmas. Consiste en explicar con codigo, como llegar a la solucion de un problema

```cpp
#include<iostream>
using namespace std;


int main(){

    // Calculadora de IMC
    float peso, altura, resultado;

    // Entrada y salida de datos
    count<<"Ingrese su peso (kgs): ";
    cin>>peso;

    count<<"Ingrese su altura (mts): ";
    cin>>altura;

    // Calculo del IMC
    resultado = peso / (altura*altura);

    // Imprimimos el resultado
    cout<<"Su IMC es: "<<resultado<<endl;

}
```

**Programacion estructurada**

Este paradigma se centra en mejorar la claridad y calidad del codigo utilizando solo funciones y tres estructuras:

* secuenciales
* condicionales
* iterativas (repetitivas)

Es una ampliacion del paradigma **imperativo**

Veamos un ejemplo:

```python
def get_lista_productos(lista):
  nombres = []
  for item in lista:
    nombres.append(item["producto"])
  return nombres


lista_compras = [
  {"producto" : "pan", "precio": 150},
  {"producto" : "arroz", "precio": 120},
  {"producto" : "milanesas", "precio": 300}
]

for item in lista_compras:
  if item["producto"] == "arroz":
    item["precio"] *= 0.75

nombres_productos = get_lista_productos(lista_compras)
print(nombres_productos)
```

Como podemos ver, con tan solo 1 funcion y las 3 estructuras, pudimos construir un programa capaz de gestionar nuestra lista de compras.

**Programacion declarativa**

A diferencia de la programacion **imperativa**, aca se prioriza el resultado antes que el paso a paso de como obtenerlo.

SQL es un ejemplo de lenguaje que utiliza este principio

```sql
SELECT * FROM usuarios WHERE nombre = 'Pedro' AND nacionalidad = 'Argentina'
```

Si analizamos un poco el ejemplo, en este lenguaje jamas le estamos indicando como conseguir los datos, solo le indicamos el resultado que esperamos: todos los usuarios de la Base de Datos que se llamen Pedro y sean argentinos.

**Programacion funcional**

En este paradigma, las funciones son tratadas como ciudadanos de primera clase. Se prioriza el uso de recursividad y **Higher-order Function** (funciones que reciben y/o devuelven funciones) para resolver problemas

```javascript
let meetups = [
    {name: 'Javascript', isActive:true, members: 700},
    {name: 'Angular', isActive:false, members: 700},
    {name: 'Node', isActive:false, members: 900},
    {name: 'React', isActive:true, members: 600},
    {name: 'Python', isActive:true, members: 1200}
];

let sum = meetups.filter(meetup => {
    return meetup.isActive;
})
.map(meetup => {
    return meetup.members - (0.01*meetup.members)
})
.reduce((acc, meetup) => {
    return acc + meetup; 
}, 0)

console.log(sum)
```

### Relaciones entre clases

**Asociacion**

La asociación es la relación entre dos clases separadas que se establece a través de sus Objetos. La asociación puede ser de uno a uno, de uno a varios, de varios a uno, de varios a varios. 

En la programación orientada a objetos, un objeto se comunica con otro objeto para utilizar la funcionalidad y los servicios proporcionados por ese objeto. La composición y la agregación son las dos formas de asociación. 

**Agregacion**

Es una forma especial de Asociación donde:  

+ Se representa la relación "tiene un" (*has a* en ingles) .
+ Es una asociación unidireccional, lo que implica que solo una de laspartes estara relacionada con la otra. Por ejemplo, un departamento puede tener estudiantes, pero viceversa no es posible y, por lo tanto, es de naturaleza unidireccional.
+ En **Agregación**, ambas clases pueden sobrevivir individualmente, lo que significa que si eliminamos una de las entidades, la otra entidad no se veria afectada.

Este ultimo punto implica que estamos ante una relacion de dependencia DEBIL.

La composicion es similar a la agregacion pero es mas restrictiva ya que en esta relacion, las clases son dependientes entre si.

class NombreDeLaClase:

### Crear objetos

Como sabemos, los objetos son instancias de clases. Los objetos son lo que realmente interactua en la aplicacion. Las clases son mas bien plantillas que definen el conjunto de metodos y atributos que tendran en comun todas sus instancias


```python
class Persona:
  pass

persona_1 = Persona()
```

Es importante prestar atencion a lo siguiente: despues de llamar a nuestra clase, debemos colocar los parentesis.

Cuando llamamos a una clase de esta forma, lo que estamos haciendo es invocar a un metodo especial de todas las clases, llamado **contructor**.

Como su nombre lo indica, este metodo "construye". Pero que? La respuesta es un poco obvia. Este metodo construye un objeto y lo devuelve

Ahora **persona_1** esta haciendo referencia a un objeto del tipo `Persona`

````python
class Persona:
  pass

persona_1 = Persona()

print(type(persona_1))
````

### Agregar atributos

Los atributos, como dijimos, son valores dentro de nuestro objeto. Algo asi como sus caracteristicas.

Si pensamos en las caracteristicas de una Persona, podemos pensar en: nombre, apellido, documento, edad, etc. Todos esos valores, son atributos.

Para agregar atributos a nuestra clase, vamos a utilizar un metodo especial: `__init__()`

Este metodo, es el constructor y es a quien debemos decirle cuales seran los atributos del objeto y cuales seran sus valores iniciales.

````python
class Persona:
  
  def __init__(self, nombre, apellido, documento, edad):
    self.nombre = nombre
    self.apellido = apellido
    self.documento = documento
    self.edad = edad

# Ahora debemos pasarle esos datos si o si al contructor de la clase
persona_1 = Persona("Pedro", "Gomez", 40898677, 22)
````

### Agregar Metodos

Los metodos dentro de las clases se definen como funciones. Las mismas que venis usando hasta ahora.

````python
class Persona:
  
  def __init__(self, nombre, apellido, documento, edad):
    self.nombre = nombre
    self.apellido = apellido
    self.documento = documento
    self.edad = edad

  def imprimir_datos(self):
    print(f"{self.apellido.upper()}, {self.nombre.title} Documento: {self.documento} - Edad: {self.edad}")

  
# Ahora debemos pasarle esos datos si o si al contructor de la clase
persona_1 = Persona("Pedro", "Gomez", 40898677, 22)
````

### Accediendo a metodos y atributos

````python
class Persona:
  
  def __init__(self, nombre, apellido, documento, edad):
    self.nombre = nombre
    self.apellido = apellido
    self.documento = documento
    self.edad = edad

  def imprimir_datos(self):
    print(f"{self.apellido.upper()}, {self.nombre.title()} Documento: {self.documento} - Edad: {self.edad}")

  
persona_1 = Persona("Pedro", "Gomez", 40898677, 22)

# Acceder a los atributos
print(f"Atributo nombre: {persona_1.nombre}")
print(f"Atributo apellido: {persona_1.apellido}")
print(f"Atributo documento: {persona_1.documento}")

# Ejecutar un metodo
persona_1.imprimir_datos()
````

Algo que seguro te llamo la atencion es ese parametro llamado `self` que jamas le pasamos.

El parámetro `self` se refiere al objeto instanciado de esa clase sobre el cual se está invocando dicho método. Es decir, el objeto que usaste para llamar al método. En nuestro caso, self seria el objeto que es apuntado por `persona_1`

Esto es asi debido a que los objetos no contienen una copia de los metodos. Seria malgastar memoria dado que los metodos no cambian. Asi que Python utiliza las referencias al objeto para ejecutar los metodos y acceder a los valores asociados a ese objeto.

En resumen, los objetos contienen sus propios atributos pero los metodos son de las clases, es decir, se comparten con todos los objetos. es por este motivo que los metodos necesitan una referencia al objeto que los esta llamando, para acceder a sus atributos.

````python

class Auto:

  def __init__(self, marca, modelo):
    self.marca = marca
    self.modelo = modelo

  def encender(self):
    print(f"Encendiendo {self.marca}")
  
  def apagar(self):
    print(f"Apagando {self.marca}")


auto_1 = Auto("Ford", "Raptor")
auto_2 = Auto("Mitsubishi", "Lancer EVO X")

auto_1.encender()
auto_2.encender()
````

Cuando invocamos el metodo desde la instancia de la clase, Python se encarga de hacer que self contenga el objeto en cuestion (`auto_1` o `auto_2`).

Un equivalente de ejecucion de este metodo seria el siguiente

````python
class Auto:

  def __init__(self, marca, modelo):
    self.marca = marca
    self.modelo = modelo

  def encender(self):
    print(f"Encendiendo {self.marca}")
  
  def apagar(self):
    print(f"Apagando {self.marca}")


auto_1 = Auto("Ford", "Raptor")
auto_2 = Auto("Mitsubishi", "Lancer EVO X")

Auto.encender(auto_1)
Auto.encender(auto_2)
````
