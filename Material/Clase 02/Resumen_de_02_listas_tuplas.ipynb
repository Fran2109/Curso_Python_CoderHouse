## Listas

Las listas son un tipo de dato compuesto. Que quiere decir eso? 

Las listas por si mismas son un tipo de dato, como `int()`. Pero a diferencia de este ultimo, que solo puede guardar un solo valor, las listas almacenan muchos valores.

La forma de declarar listas es definir todos sus elementos entre corchetes y separados por coma:

```python
mi_lista_de_calificaciones = [6.50, 9.70, 10]
heroes_marvel = ["Iron Man", "Dr. Strange", "Hulk", "Captain America"]
```

A diferencia de otros lenguajes de programacion, en Python las listas son Heterogeneas. Esto quiere decir que podemos tener distintos tipos de datos en una misma lista

```python
variable = 123
datos_varios = [2022, 32.5, variable, "Esto es una cadena"]
```

Ahora que ya tenemos datos en nuestras listas, talvez lo mas interesante seria poder acceder a ellos. Pero como?

Ya lo hemos visto con los strings: indices y corchetes

Las reglas de los indices en los strings, se aplican a cualquier coleccion de datos en Python.

```python
heroes_marvel = ["Iron Man", "Dr. Strange", "Hulk", "Captain America", "Black Widow", "Black Panther"]


# Primer heroe
print(heroes_marvel[0])

# Ultimo heroe
print(heroes_marvel[-1])
```

Tambien, como sucede con los strings, si sumamos dos listas obtendremos como resultado una nueva lista con los elementos de la lista1 y la lista2.

Veamos

```python
lista1 = [1,2,3,4]
lista2 = [4,5,6,7,8]

lista_nueva = lista1 + lista2

print(lista_nueva)
```

Te acordas que los strings eran INMUTABLES? Bueno, con las listas eso no pasa.

Podemos cambiar los valores en los indices que queramos

```python
mi_lista_de_calificaciones = [6.50, 9.70, 10]

mi_lista_de_calificaciones[2] = 8.66

print(mi_lista_de_calificaciones)
```

Ahora que sabemos que son MUTABLES, podemos modificarlas sin problemas. Como por ejemplo en la siguiente situacion:

Me equivoque cargando heroes de Marvel, agregando accidentalmente heroes de DC. Necesito borrar esos valores.

Que puedo hacer? Borrarlos mediante Slicing es una opcion

```python
heroes_marvel = ["Iron Man", "Dr. Strange", "Hulk", "Captain America", "Black Widow", "Batman", "Superman"]

# Al asignarle a esa porcion de la lista, una lista vacia, los elementos se borran.
heroes_marvel[5:] = []

print(heroes_marvel)

# Si queremos vaciar la lista, hacemos esto
heroes_marvel = []

print(heroes_marvel)
```

## Funciones de listas

Tenemos varias funciones integradas en el tipo de dato "list". En esta ocasion veremos las mas importantes: append(), pop(), count(), e index().

La funcion `append()` nos permite agregar elementos al final de la lista.

### Obtener la longitud de una lista

`len()` no solo recibe strings. Esta funcion recibe cualquier coleccion, por ende, puede calcular la longitud de una lista

```python
numeros_loteria = [101, 100, 156, 248]

print("Cantidad de numeros obtenidos: ", len(numeros_loteria))
```

### Quitar elementos

No solo el slicing me permite quitar elementos de una lista.
La funcion `pop()` recibe opcionalmente un argumento: la posicion del elemento a borrar.

Si no lo recibe, borrara el ultimo elemento. [Demostracion](https://tenor.com/bPQv4.gif)

Si lo recibe, buscara ese indice y lo destruira. [Demostracion](https://tenor.com/view/python-pop-index-gif-24710585)

`pop()` eliminara ese elemento de la lista, reacomodara los elementos y los indices, y devolvera el elemento que fue quitado.

```python
# Para que sea mas comodo trabajar, podemos acomodar la lista de la siguiente forma
# Al encontrar la coma, Python sabe que hay una continuacion de elementos
# Sino la encuentra despues de la coma, buscara en la siguiente linea. Esto hasta encontrar
# el cierre "]"
heroes_marvel = ["Iron Man", 
                 "Dr. Strange", 
                 "Hulk", 
                 "Captain America", 
                 "Black Widow", 
                 "Black Panther"]


# Eliminar ultimo heroe
heroe = heroes_marvel.pop()
print("Hemos eliminado a", heroe, ". Quedan ", heroes_marvel)

# Eliminar un heroe en un indice
heroe = heroes_marvel.pop(0)
print("Hemos eliminado a", heroe, ". Quedan ", heroes_marvel)

# Eliminar un heroe en un indice que el usuario indique
cantidad_heroes = len(heroes_marvel)
mensaje = "Indique el indice a eliminar. Valores entre 0 y " + str(cantidad_heroes) + ": "

indice = int(input(mensaje))
heroe = heroes_marvel.pop(indice)

print("Hemos eliminado a", heroe, ". Quedan ", heroes_marvel)
```

### Count

`count()` es una funcion interesante.
Recibe como argumento un valor, y recorrera toda la lista buscando lo que hayamos indicado y contando la cantidad de veces que lo encontro.

Una vez que se recorrio toda la lista, nos devolvera un valor numerico entero que representa la cantidad de veces que es valor esta repetido.

Te dejo este GIF, para que entiendas mejor: [Demostracion](https://giphy.com/gifs/hNupwh83TP40pYOVkc/fullscreen)

```python
calificaciones = [82, 87, 57, 85, 2, 45, 25, 62, 20, 57]

print("Cantidad de alumnos que sacaron 57: ", calificaciones.count(57))

# Elemento que no existe
print("Cantidad de alumnos que sacaron 1: ", calificaciones.count(1))
```

### Index

`index()` es una funcion que recibe como argumento un valor (similar a `count()`) y devolvera un numero entero que representa el indice donde se encuentra el valor que estamos buscando.

Esta funcion, a diferencia de `count()`, si le indicamos un valor que no existe en nuestro programa, simplemente fallara.

## Tuplas

Las tuplas son colecciones de datos muy similares a las listas, pero con una diferencia importante: son INMUTABLES!

Para que querria yo una coleccion de datos que no puedo cambiar? Esta coleccion es muy util cuando necesitamos "agrupar" varios datos en un mismo lugar, pero evitando que se modifiquen en algun momento.

Imaginemos que necesitamos almacenar el nombre de usuario, email y contrasea de un usuario en especifico. Con una lista, alguien podria hacer que el nombre del usuario cambie, por ejemplo.

Las tuplas se definen de forma similar a las listas, pero en lugar de usar corchetes para AGRUPAR los elementos, lo hacemos con parentesis

```python
# Tupla de varios elementos
tupla = (1,2,3,"Hola mundo")

print("Contenido de la tupla:", tupla, "Tipo de dato:", type(tupla))


# Tupla de 1 elemento
# Cuando se trata de un solo elemento, debemos de agregar una coma al final para que 
# Python sepa que se trata de la definicion de una tupla
tupla_simple = (1,)

print("Contenido de la tupla:", tupla_simple, "Tipo de dato:", type(tupla_simple))


# De lo contrario, creera que es un entero
tupla_simple = (1)

print("Contenido de la tupla:", tupla_simple, "Tipo de dato:", type(tupla_simple))
```

Igual que en las listas:
* Las tuplas tambien son HETEROGENEAS
* Podemos acceder a sus elementos por medio de indices
* Podemos aplicar slicing

```python
# Tupla heterogenea
usuario = ("usuario_admin", "admin@example.com", "password123", 2022)

# Accedemos al primer indice para mostrar el nombre de usuario
print("Nombre de usuario:", usuario[0])

# Por medio de slicing, mostramos el correo y la contrasea
print("Datos para iniciar sesion:", usuario[1:3])
```

### Agregar elementos

Dado que no son mutables, no podemos hacer uso de `append()` ni `pop()`, pero si podemos sumar tuplas (concatenarlas).


```python
tupla_1 = (1,2,3,"Hola mundo")

tupla_1 = tupla_1 + (2,3,4)

print("Elementos de la tupla:", tupla_1)
```

Como es posible que podamos agregar mas valores concatenando otra tupla y no podamos con un `append()`?

Internamente, lo que hicimos fue volver a definir el contenido de la variable tupla_1.

Python resuelve la suma de tuplas (que da como resultado una nueva tupla), crea el objeto en cuestion y hace la asociacion variable-objeto.

En sintesis, la variable apunta a una nueva tupla.

Si vemos el siguiente ejemplo con las listas, el objeto lista que es referenciado por la variable `lista`, sigue siendo el mismo ya que el ID no cambia

```python
lista = [1,2,3]

print("ID de la lista: ", id(lista), "Elementos", lista)

lista.append(4)
print("ID de la lista: ", id(lista), "Elementos", lista)

lista.append(5)
print("ID de la lista: ", id(lista), "Elementos", lista)

lista.append(6)
print("ID de la lista: ", id(lista), "Elementos", lista)
```

Si aplicaramos una concatenacion de listas, la cosa cambia

```python
lista = [1,2,3]

print("ID de la lista: ", id(lista), "Elementos", lista)

lista = lista + [4,5,6]

print("ID de la lista: ", id(lista), "Elementos", lista)
```

Como podemos ver, el ID ha cambiado. Lo que significa que la variable "lista" no esta referenciando al mismo objeto de antes.

Esto se debe a que cada vez que usamos el operador de asignacion "=", Python crea un objeto y lo referencia con la etiqueta que indicamos.

Lo mismo pasa con las tuplas.

```python
tupla = (1,2,3)

print("ID de la tupla: ", id(tupla), "Elementos", tupla)

tupla = tupla + (4,5,6)

print("ID de la tupla: ", id(tupla), "Elementos", tupla)
```

### Borrar elementos de una tupla

El borrado por slicing no se puede realizar por obvias razones (no podemos modificar los indices), pero si le podemos decir a la nuestra variable que apunte a una tupla vacia

```python
tupla = (1,2,3)

print("ID de la tupla: ", id(tupla), "Elementos", tupla)

tupla = ()

print("ID de la tupla: ", id(tupla), "Elementos", tupla)
```

### Funciones y Funciones integradas

Las funciones integradas dentro de una tupla son parecidas a la de las listas

```python
tupla = tupla = (1,2,3)

# Longitud de una tupla
print( len(tupla) )

# Cantidad de veces que un elemento se encuentra dentro de la tupla
print( tupla.count(1) )

# Obtener el indice de un elemento
print( tupla.index(2) )
```
