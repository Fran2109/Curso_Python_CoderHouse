## Operadores

Un operador es un smbolo o palabra, que indica que debe ser llevada a cabo una operacin especificada sobre un cierto nmero de operandos.

Nosotros ya hemos conocido algunos de ellos: +, -, * /, etc.

Los operadores nos permiten formar "expresiones":

Operando1 [Operador] Operando2

En Python, el comportamiento del Operador depende de los tipos de datos de los operandos

En este caso `2 + 1`, el operador SUMA resolvera una operacion matematica. Sin embargo, en este caso `"Hola"+" Mundo"` el operador SUMA hara una concatenacion (lo mismo pasa con listas y tuplas)

Las expresiones que emplean operadores aritmeticos, se dividen en 2 partes:

* Expresiones aritmeticas: son expresiones donde ambos operandos, son literales: `2+5`, `-1.4*54`
* Expresiones algebraicas: son expresiones donde al menos uno de los operandos, es una variable: `radio*3.14`

## Tipo de dato Logico

El tipo de dato logico, tambien conocido como tipo Booleano o Binario, es el mas basico de todos.

Solo admite dos valores: Verdadero o Falso.

En la vida, nos la pasamos evaluando la veracidad de las oraciones.

"Hoy es Lunes", "El cielo es morado", "1+1 es 3", etc.

Todas estas oraciones tiene algo en comun: o son Verdaderas o son Falsas.

En el mundo de la logica booleana, la negacion de algo Verdadero lo convierte en Falso. Y la negacion de algo Falso, lo convierte en algo Verdadero.



## Operadores relacionales

Estos operadores tambien llevan adelante una operacion, pero en este caso, es la de COMPARACION.

Si! Asi como lo lees. Ahora vamos a comparar Operandos.

El resultado de comparar algo, da como resultado un valor Booleano (Verdadero o Falso).

A las expresiones formadas por estos operadores, se les conoce como "expresiones booleanas" o "expresiones logicas"

## Operadores logicos

Los operadores logicos nos permiten crear GRANDES expresiones booleanas, uniendo simples expresiones logicas.

En Python existen 3:

* and
* not
* or

### Operador NOT

Este operador representa la Negacion Logica o Booleana.

"Si negamos algo que es Verdadero, se convierte en Falso"

Con este operador, aplicamos ese concepto

````python
print( not True )
print( not True )

# Primero se evalua el operador relacional ==, que da como resultado True
# Y luego se resuelve el operador logico not, que da como resultado False
print( not 5 == 5 )
````

### Operador AND

Es un operador de conjuncion que une dos operandos. Estos operandos pueden ser expresiones logicas o valores booleanos

Si ambos operandos son Verdaderos, la expresion completa sera verdadera.

Si al menos uno de los operandos es Falso, toda la expresion sera Falsa


````python
usuario = "Admin"
edad = 23

print( usuario == "Admin" and edad > 18)

# En este caso, la expresion usuario=="admin" es Falsa, pero edad > 18 es Verdadera
# La definicion de AND dice que "todos los operandos deben ser verdaderos para que la expresion sea verdadera"
# Como esto no es asi, la expresion da como resultado un False.
print( usuario == "admin" and edad > 18)
````

### Operador OR

Es un operador de disyuncion, y como su nombre lo indica, separa.

A diferencia del AND, el OR necesita que al menos 1 de los operandos sea Verdadero para declarar la expresion completa como tal.

````python
edad = 25
pais = "Argentina"

print( pais == "Argentina" or edad > 30 )

# Si ambos operandos son Falsos, la expresion completa dara ese resultado
print( pais != "Argentina" or edad > 30 ) 
````

## Expresiones anidadas

Ahora que conocemos muchos operadores, podemos armar expresiones tan complejas como se nos ocurra.

Lo que si tenemos que tener en cuenta es el orden de procedencia

Python resuelve las expresiones en el siguiente orden

1. Operadores aritmeticos
    
    1. Terminos entre parentesis
    2. Potencias y raices
    3. Multiplicaciones y divisiones
    4. Sumas y Restas
2. Operadores relacionales: todos los operadores tienen la misma jerarquia por lo que se resuelve de izquiera a derecha
3. Operadores logicos:

  1. not
  2. and y or: tiene la misma jerarquia por lo que se resuelve de izquiera a derecha

````python
a = 5
b = 3

resultado = a ** b / 3 + 5 > 10 and not (a%b) == 0
# Paso a paso
#            5**3  / 3 + 5 > 10 and not (5%3) == 0      Reemplazamos los variables por sus valores
#            5**3  / 3 + 5 > 10 and not (2)   == 0      Resolvemos terminos entre parentesis
#            125   / 3 + 5 > 10 and not (2)   == 0      Resolvemos potencias      
#   41.666666666666664 + 5 > 10 and not (2)   == 0      Resolvemos divisiones
#       45.666666666666664 > 10 and not (2)   == 0      Resolvemos las sumas
#                          True and not (2)   == 0      Resolvemos los operadores relacionales de izq. a der
#                          True and not False           Resolvemos el operador logico NOT
#                          True and True                Resolvemos el operador AND
#                              True                     Resultado final
print(resultado)
````

## Operadores de Asignacion

Los operadores de asignacion operadores aritmeticos usados para modificar el valor actual de una variable.

El operador de asignacion mas utilizado es el signo "="

Pero no es el unico.

Teniendo ya declarada la variable, podemos sumarle un valor numerico:



```
a = 0
a += 1
```

Ahora la variable "a" tendra el valor de 1.

Solo sirve con sumas? NO! Podemos usar cualquier operador aritmetico


````python
a = 50

# Suma
a += 1
print(a)

# Resta
a -= 1
print(a)

# Multiplicacion
a *= 25
print(a)

# Division
a /= 2
print(a)

# Division entera
a //= 2
print(a)

# Potencia
a **= 2
print(a)

# Resto
a %= 2
print(a)
````
